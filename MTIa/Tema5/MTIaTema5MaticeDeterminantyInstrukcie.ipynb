# <font color=brown> Motivácia </font>

**Pri návrhu novej generácie komerčných a vojenských lietadiel** používajú inžinieri v spoločnosti Boeing Phantom Works 3D modelovanie a výpočtovú dynamiku tekutín (CFD). Študujú prúd vzduchu okolo virtuálneho lietadla, aby pred vytvorením fyzických modelov odpovedali na dôležité otázky týkajúce sa návrhu. Tým sa drasticky znížili časy i náklady na návrh a **lineárna algebra** zohráva v tomto procese zásadnú úlohu.

Virtuálne **lietadlo** začína ako **matematický „drôtový model“**, ktorý existuje iba v počítačovej pamäti a na grafických zobrazovacích termináloch (Viď Boeing 777 na obrázku). Tento matematický model ovplyvňuje každý krok navrhovania a výroby letúna - exteriéru aj interiéru. Analýza CFD sa týka vonkajšieho povrchu. 

Aj keď sa konečná povrchová plocha roviny môže javiť ako hladká, geometria povrchu je komplikovaná. Lietadlo má okrem krídel a trupu aj gondoly, stabilizátory, laty, klapky a krídla. Spôsob prúdenia vzduchu okolo týchto štruktúr určuje, ako sa lietadlo pohybuje po oblohe. Rovnice, ktoré opisujú prúdenie vzduchu, sú zložité a musia zodpovedať nasávaniu motora, výfukom motora a brázdám, ktoré zanechávajú krídla lietadla. Na štúdium prúdenia vzduchu potrebujú inžinieri vysoko prepracovaný popis povrchu lietadla.

**Modelovanie povrchu lietadla** pomocou počítača je netriviálny proces, vyžadujúci si pokročilé techniky napr. **lineárnej algebry**, diferenciálnej geometrie či parciálnych diferenciálnych rovníc. Kľúčovým prvkom pri navrhovaní lietadiel, konkrétne pri procese skúmania prúdenia vzduchu okolo lietadla zahŕňa opakované riešenie extrémne veľkých **sústav lineárnych rovníc**, kde vstupujú do hry okrem iného aj špeciálne (blokové) **matice**. Pri numerických výpočtoch pomáhajú i rôzne **rozklady matíc**. 

Aby bolo možné analyzovať riešenie systému prúdenia vzduchu, potrebujú inžinieri vizualizovať prúdenie vzduchu po povrchu lietadla. Používajú počítačovú grafiku a **lineárna algebra** je základom takej grafiky. Drôtový model povrchu lietadla je uložený ako **dáta** v mnohých **maticiach**. Po vykreslení obrázka na obrazovke počítača môžu inžinieri zmeniť mierku, priblížiť alebo oddialiť malé oblasti a otočiť obrázok, aby videli časti, ktoré môžu byť skryté z určitého pohľadu. Každá z týchto operácií sa vykonáva vhodným **násobením** vhodných **matíc**.

V téme 6 si teda približíme koncepty **matíc** a **determinantov**. Zavedieme rôzne **typy matíc** a ukážeme si ako funguje **maticová aritmetika** (operácie s maticami). Na príkladoch budeme ilustrovať **vlastnosti** týchto **operácii**. Pozrieme sa na to, ako sa dopracovať k pojmu **determinant matice**, ukážeme si ako ho vypočítať a aké rôzne **vlastnosti** platia pre **determinanty matíc**. Spomenieme aj vzťah **inverznej matice** a **determinantu**. Ako neskôr uvidíme, **inverzná matica** i **determinant** budú užitočné okrem iného aj pri **riešení sústav lineárnych rovníc**.
