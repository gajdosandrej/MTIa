Uvažujme na chvíľu matice s jedným stĺpcom - teda **stĺpcové vektory** (analogické úvahy je možné robiť aj pre riadkové vektory) nad poľom reálnych čísel. Konkrétne nás budú zaujímať vektory s dvoma a troma zložkami lebo ako uvidíme, tieto si vieme jednoducho znázorniť.  Majme napr. vektor $\mathbf{a}=
\begin{pmatrix}
2 \\
1
\end{pmatrix}$. Vo fyzike a niekedy aj v geometrii sa zvykne vektor označovať šípkou t.j. $\vec{\mathbf{a}}=
\begin{pmatrix}
2 \\
1
\end{pmatrix}.$ Tento vektor máme znázornený na obrázku nižšie. Začiatok vektora (počiatočný bod) sme položili do bodu $[0,0]$ (stred resp. počiatok súradnicovej sústavy) a koncový bod je v bode $[2,1]$ (posun o $2$ vpravo v smere $x$-ovej osi a posun o 1 nahor v smere osi $y$). Každý vektor (jeho počiatočný bod) je možné presunúť do stredu (počiatku) súradnicovej sústavy t.j. do bodu $[0,0]$. 

Vektory ako špeciálne prípady matíc sa zvyknú označovať malými (tučnými/hrubými) písmenami. Miesto stĺpcového vektora $\begin{pmatrix}
2 \\
1
\end{pmatrix}$ sa zvykne niekedy používať **úspornejší ale ekvivalentný** (vyjadrujúci ten istý stĺpcový vektor) **zápis** pomocou tzv. transpozície v tvare $(2,1)^{T}$ alebo $(2,1)'$. Dokonca ako pri maticiach, môžete sa stretnúť aj s hranatými zátvorkami v zápise vektora $\begin{bmatrix}
2 \\
1
\end{bmatrix}$. Ak by sme v našom príklade pomenovali začiatočný a koncový bod vektora napr. $A=[0,0]$, $B=[2,1]$, potom by sme vektor (rozdiel súradníc koncového a začiatočného bodu) mohli označiť aj $\vec{AB}=B-A=[2,1]-[0,0]=(2-0,1-0)^{T}=(2,1)^{T}$.  

```python
# vektor (2,1)
# figsize parameter sluzi na regulaciu velkosti vykresleneho obrazka  
arrow((0,0), (2,1), figsize=3)
```

#### Poznámka 1

* Ak vynásobime vektor skalárom, ktorý je rovný $1$, pôvodný vektor sa nezmení. 


* Ak vynásobime vektor skalárom, ktorý je rovný $0$, pôvodný vektor sa zmení na **nulový vektor**, t.j. $(0,0)^{T}$ resp. $(0,0,0)^{T}$. 


* Po vynásobení vektora záporným číslom sa zmení vektor na opačný (šípka opačným smerom ako pôvodný vektor). 


* Majme daný vektor $\mathbf{v}=(a,b)^{T}$, jeho **dĺžku (veľkosť)** vypočítame z pravouhlého trojuholníka aplikovaním Pythagorovej vety tak, ako je to znázornené na obrázku nižšie, teda $||\mathbf{v}||=\sqrt{a^2+b^2}$, kde $||\mathbf{v}||$ označuje **dĺžku (veľkosť resp. normu) vektora** $\mathbf{v}$. V prípade trojrozmerného vektora $\mathbf{v}$ so súradnicami (zložkami) $v_1,v_2,v_3$ t.j. $\mathbf{v}=(v_1,v_2,v_3)^{T}$ by bola jeho veľkosť daná vzťahom $||\mathbf{v}||=\sqrt{v_1^2+v_2^2+v_3^2}$ (pre názornosť, stačí si nakresliť napr. uhlopriečku v kvádri s dĺžkami strán $a,b,c$ a dvakrát postupne použiť Pythagorovu vetu na výpočet dĺžky tejto uhlopriečky - viď obrázok nižšie pre vzdialenosť dvoch bodov v 3D). 


* Ak je vektor vynásobený číslom $p$ tak dĺžka (veľkosť) vektora sa zmení $|p|$-násobne, kde $|p|$ označuje absolútnu hodnotu čísla $p$. Inými slovami **dĺžka (veľkosť)  vynásobeného vektora bude rovná $|p|$ krát dĺžka (veľkosť) pôvodného vektora**. Formálne to môžeme zapísať takto: Majme daný vektor $\mathbf{v}$ a skalár $p$; nech dĺžka (veľkosť) vektora $\mathbf{v}$ je označená symbolom $||\mathbf{v}||$; potom $p$-násobok vektora $\mathbf{v}$ t.j. vektor $p\mathbf{v}$ má dĺžku (veľkosť) $|p|\cdot||\mathbf{v}||$, tzn. $||p\mathbf{v}||=|p|\cdot||\mathbf{v}||$. 



$$\text{Zdroj: Poole (2014).}$$

#### Poznámka 2

* Uvedomme si (viď obrázky), že odčítať jeden vektor od druhého vlastne znamená pripočítať vektor opačný. 


* Všetky prezentované poznatky pre vektory v 2D a 3D je možné analogicky zovšeobecniť pre viacrozmerné vektory, vo všeobecnosti pre vektory s $n$ zložkami, kde $n\in\mathbb{N}$. Vo viacerých rozmeroch je však už vizualizácia obtiažna resp. nemožná. 


* Majme teda vektory $\mathbf{a}=(a_1,a_2,\ldots,a_n)^{T}$, $\mathbf{b}=(b_1,b_2,\ldots,b_n)^{T}$ také, že majú $n$ zložiek (prvkov) a tieto ich zložky (prvky) sú reálne čísla, skrátene to zapisujeme $\mathbf{a},\mathbf{b}\in\mathbb{R}^{n}$, čiže máme v podstate dve $n$-prvkové usporiadané postupnosti reálnych čísel - tzv. $n$-tice. Nech $p$ je prvok z poľa, v tomto prípade nech je to nejaké reálne číslo. Operácie s dvoma $n$-rozmernými vektormi môžeme zapísať vo všeobecnosti takto: 

$$\mathbf{a}+\mathbf{b}=\left(a_1+b_1,\ldots,a_n+b_n\right)^{T},$$

$$\mathbf{a}-\mathbf{b}=\left(a_1-b_1,\ldots,a_n-b_n\right)^{T},$$

$$p\cdot\mathbf{a}=\left(p\cdot a_1,\ldots,p\cdot a_n\right)^{T}.$$

#### Poznámka 3

* Všimnime si, že pri operáciach s vektormi, ktoré sme zatiaľ prezentovali, bol vždy dvom zvoleným vektorom priradený tretí (výsledný) vektor ako výsledok operácie týchto dvoch vektorov. Z násobenia matíc už vieme, že súčin riadkového a stĺpcového vektora je skalár (prvok z poľa) - teda napr. nejaké číslo. Túto operáciu nazývame **skalárny súčin vektorov**. Na rozdiel od predošlých operácii s vektormi, pri skalárnom súčine je dvom vektorom priradený skalár (číslo) a nie vektor! 


* V téme 5 (pri súčine matíc) sme videli, že súčin dvoch matíc sú vlastne postupné súčiny riadkov prvej matice so stĺpcami druhej matice - teda skalárne súčiny vektorov. Vo všeobecnosti, keď máme dva vektory $\mathbf{a},\mathbf{b}\in\mathbb{R}^{n}$, tzn. vektory s $n$ zložkami (prvkami, kde prvky sú reálne čísla), t.j. $\mathbf{a}=(a_1,a_2,\ldots,a_n)^{T}$, $\mathbf{b}=(b_1,b_2,\ldots,b_n)^{T}$, tak ich **skalárny súčin je definovaný** vzťahom  

$$\mathbf{a}^{T}\cdot\mathbf{b}=(a_1,a_2,\ldots,a_n)\cdot\begin{pmatrix}b_1\\b_2\\ \vdots \\ b_n\end{pmatrix}=a_1b_1+a_2b_2+\ldots a_nb_n.$$ 


* Na základe predošlej odrážky môžeme dĺzku (veľkosť/normu) $n$-rozmerného vektora $\mathbf{v}\in\mathbb{R}^{n}$ vo všeobecnosti napísať takto 

$$||\mathbf{v}||=\sqrt{\mathbf{v}^{T}\cdot\mathbf{v}}=\sqrt{v_1^2+\ldots v_n^2}.$$ 


* Na konkrétnom príklade (ale aj vo všeobecnosti) je možné ľahko overiť, že ak máme ľubovoľný vektor $\mathbf{v}$ a z neho vytvoríme vektor $\mathbf{v}^*$ tak, že vynásobime pôvodný vektor $\mathbf{v}$ prevrátenou hodnotou jeho normy t.j. $\mathbf{v}^*=\dfrac{1}{||\mathbf{v}||}\cdot\mathbf{v}$, potom nový vektor $\mathbf{v}^*$ bude mať normu (resp. dĺžku) rovnú hodnote $1$ (viď obrázok). Taký vektor sa zvykne nazývať aj **_normovaný_** alebo **_jednotkový vektor_**.

 


* Niekedy sa skalárny súčin vektorov $\mathbf{a},\,\mathbf{b}$ značí iba $\mathbf{a}\cdot\mathbf{b}$ alebo $\mathbf{a}\mathbf{b}$, my však budeme používať označenie $\mathbf{a}^{T}\cdot\mathbf{b}$ alebo $\mathbf{a}^{T}\mathbf{b}$, kde horný index $^T$  znamená, že $\mathbf{a}^T$ je riadkový vektor, $\mathbf{b}$ je stĺpcový vektor, t. j. násobime "riadok krát stĺpec" - ako pri násobení matíc; transpozícia $^T$ len preklopí pôvodne stĺpcový vektor $\mathbf{a}=\begin{pmatrix}a_1\\a_2\\ \vdots \\ a_n\end{pmatrix}$ na riadkový vektor $\mathbf{a}^{T}=(a_1,a_2,\ldots,a_n)$.

#### <font color=green> Tvrdenie 1 </font>

Nech $\mathbf{u}$, $\mathbf{v}$, $\mathbf{w}$ sú vektory z $\mathbb{R}^n$ a nech $c$ je skalár. Potom platia nasledujúce vzťahy: 

* (i) $\mathbf{u}^T\cdot\mathbf{v}=\mathbf{v}^T\cdot\mathbf{u}$; 


* (ii) $\mathbf{u}^T\cdot(\mathbf{v}+\mathbf{w})=\mathbf{u}^T\cdot\mathbf{v}+\mathbf{u}^T\cdot\mathbf{w}$;


* (iii) $(c\mathbf{u})^T\cdot\mathbf{v}=c(\mathbf{u}^T\cdot\mathbf{v})$; 


* (iv) $\mathbf{u}^T\cdot\mathbf{u}\ge0$ a $\mathbf{u}^T\cdot\mathbf{u}=\mathbf{0}$ práve vtedy, keď $\mathbf{u}=\mathbf{0}$; 


* (v) $||\mathbf{v}||=0$ práve vtedy, keď $\mathbf{v}=\mathbf{0}$. 

---

<font color=orange> Na konkrétnych príkladoch ilustrujte platnosť vzťahov uvedených v Tvrdení 1 (tzn. zvoľte si ľubovoľné tri vektory $\boldsymbol{u}$, $\boldsymbol{v}$, $\boldsymbol{w}$, konštantu $c$ a rozpíšte pravú i ľavú stranu vo vzťahoch (i)-(iv)). </font>

### <font color=green> Veta 1 (Cauchyho-Schwartzova nerovnosť) </font>

Pre ľubovoľné dva vektory $\mathbf{u},\mathbf{v}\in\mathbb{R}^n$ platí 

$$|\mathbf{u}^T\cdot\mathbf{v}|\le||\mathbf{u}||\cdot||\mathbf{v}||.$$

---

<font color=orange> Na konkrétnom príklade dvoch vektorov ilustrujte nerovnosť z Vety 1. </font>

Ak si predstavíme, že vektory $\mathbf{u}$, $\mathbf{v}$ sú strany nejakého trojuholníka, tak vieme, že súčet dĺžok dvoch strán v trojuholníku musí byť väčší ako dĺžka tretej strany v tomto trojuholníku. Toto platí pre každý trojuholník. Ak teda označíme strany v trojhuolníku ako vektory $\mathbf{u}$, $\mathbf{v}$, $\mathbf{u}+\mathbf{v}$ a dĺžky strán v trojuholníku označíme $||\mathbf{u}||$, $||\mathbf{v}||$, $||\mathbf{u}+\mathbf{v}||$, tak dostávame **trojuholníkovú nerovnosť** v nasledujúcom (všeobecnom) tvare (Veta 2): 

---

### <font color=green> Veta 2 (Trojuholníková nerovnosť) </font>

Pre ľubovoľné dva vektory $\mathbf{u},\mathbf{v}\in\mathbb{R}^n$ platí 

$$||\mathbf{u}+\mathbf{v}||\le||\mathbf{u}||+||\mathbf{v}||.$$

Ďalej si vysvetlíme, **ako určiť vzdialenosť dvoch vektorov**. Poďme na to postupne a vychádzajme z toho, čo už vieme resp. z toho, čo si vieme ľahko znázorniť. <br/> 
Vzdialenosť dvoch vektorov je analógiou vzdialenosti dvoch bodov na reálnej osi resp. dvoch bodov v rovine (v 2D). Vzdialenosť dvoch bodov $a,b$ na reánej osi nám znázorňuje nasledujúci obrázok, konkrétne pre $a=-2$, $b=3$. 

 
$$\text{Zdroj: Poole (2014).}$$ 

Ich vzdialenosť $d$ (ang. *distance*) je $d=|a-b|$ (absolútna hodnota nám zabezpečuje, že nemusíme riešiť, ktoré z čísel $a,b$ je väčšie a ich vzdialenosť bude nezáporná). Vzťah $|a-b|$ je to isté, čo vzťah $\sqrt{(a-b)^2}$. Konkrétne teda vzdialenosť bodov $a=-2$, $b=3$ je rovná $d=|a-b|=|-2-3|=5$. <br/> 
Dvojrozmerná analógia tohto vzťahu je vzdialenosť $d$ dvoch bodov v rovine (v 2D) so súradnicami $[a_1,a_2]$, $[b_1,b_2]$ a tá je daná vzťahom $d=\sqrt{(a_1+b_1)^2+(a_2-b_2)^2}$ (viď ďalší obrázok). V podstate je tu použitá Pythagorova veta, pričom vzdialenosť daných dvoch bodov je dĺžka prepony v pravouhlom trojuholníku. 

 
$$\text{Zdroj: Poole (2014).}$$ 


Už možno tušíme, ako by vyzeral výpočet vzdialenosti dvoch bodov v 3D. 
Majme body $P=[x_0,y_0,z_0]$ a $Q=[x_1,y_1,z_1]$ v 3D (v trojrozmernom priestore $\mathbb{R}^3$). **Vzdialenosť** $d(P,Q)$ týchto **dvoch bodov v trojrozmernom priestore** určíme postupným dovojnásobným aplikovaním Pythagorovej vety (viď obrázok kvádra a popis nižšie) vzťahom 

$$d(P,Q)=|PQ|=\sqrt{(x_1-x_0)^2+(y_1-y_0)^2+(z_1-z_0)^2}.$$ 

 
$$\text{Zdroj: Schlicker et al. (2018).}$$

Najprv sa pomocou Pythagorovej vety vypočíta dĺžka $|PR|$ (červená úsečka) a to z pravouhlého trojuholníka $PSR$ s pravým uhlom pri vrchole $S$. Následne sa aplikovaním Pythagorovej vety určí požadovaná dĺžka $|PQ|$ a to z pravouhlého trojuholníka $PRQ$ s pravým ulom pri vrchole $R$. Takto dostaneme vzorec pre vzdialenosť bodov $P$ a $Q$ (dĺžka modrej úsečky) uvedený vyššie. <br/>
Analogicky by vyzeral vzorček pre výpočet vzdialenosti dvoch bodov v $n$-rozmernom priestore (pod odmocninou bude vo všeobecnosti $n$ členov resp. sčítancov) ... 

Odrazme sa ďalej od vzorčeka pre vzdialenosť dvoch bodov v rovine. Majme dva vektory $\mathbf{a}$, $\mathbf{b}$, pričom obidva nech začínajú v bode $[0,0]$ a nech majú koncové body so súradnicami $[a_1,a_2]$ a $[b_1,b_2]$ tak, ako to zobrazuje nasledujúci obrázok. 

 
$$\text{Zdroj: Poole (2014).}$$ 

Potom vzdialenosť $d$ uvažovaných dvoch vektorov $\mathbf{a}$, $\mathbf{b}$ je vlastne vzdialenosť ich koncových bodov $[a_1,a_2]$, $[b_1,b_2]$, resp. dĺžka vektora $\mathbf{a}-\mathbf{b}$. Formálne to zapíšeme takto: $d=\sqrt{(a_1-b_1)^2+(a_2-b_2)^2}=||\mathbf{a}-\mathbf{b}||$.

### <font color=red> Definícia 1 </font>

Vzdialenosť $\mathrm{d}(\mathbf{u},\mathbf{v})$ medzi dvoma vektormi $\mathbf{u},\mathbf{v}\in\mathbb{R}^n$ je definovaná vzťahom 

$$\mathrm{d}(\mathbf{u},\mathbf{v})=||\mathbf{u}-\mathbf{v}||.$$

---

<font color=orange> Pomocou Sage znázornite vektory $\mathbf{u}=(\sqrt{2},1,-1)^T$, $\mathbf{v}=(0,2,-2)^T$ a tiež vektor $\mathbf{u}-\mathbf{v}$, ktorého dĺžka vyjadruje vzdialenosť medzi vektormi $\mathbf{u}$, $\mathbf{v}$. Vypočítajte vzdialenosť daných vektorov $\mathbf{u},\mathbf{v}$. </font>

Ďalší pojem, ktorý úzko súvisí so skalárnym súčinom vektorov (i keď to zatiaľ nemusí byť vidieť) je **odchýlka dvoch  vektorov**. Niekoľko príkladov odchýliek medzi dvoma vektormi $\boldsymbol{u}$, $\boldsymbol{v}$ máme znázornených na nasledujúcich obrázkoch. 

 
$$\text{Zdroj: Poole (2014).}$$ 

Za odchýlku $\theta$ medzi dvoma vektormi budeme považovať číslo z intervalu $\langle0,180^\circ\rangle$. Uvažujme trojuholník pozostávajúci z vektorov $\mathbf{u}$, $\mathbf{v}$,  $\mathbf{u}-\mathbf{v}$, pričom vektory $\mathbf{u}$, $\mathbf{v}$ zvierajú uhol $\theta$, ako to ilustruje obrázok nižšie. 


$$\text{Zdroj: Poole (2014).}$$  

Jedna z možností ako sa dopracovať k vyjadreniu kosínusu odchýlky $\theta$ dvoch vektorov $\mathbf{u}$, $\mathbf{v}$ je nasledovná úvaha: Aplikujme [kosínusovú vetu](https://www.youtube.com/watch?v=HtVKMYeJjpM) (pozri tiež [odkaz 2](https://www.galeje.sk/web_object/9480.pdf) alebo [odkaz 3](https://cloud7.edupage.org/cloud/M210_Sinusova_a_kosinusova_veta.pdf?z%3AJa%2BD7BEp08xe1YMyVFMuCAD7Ds389MLraJgOjG2YqcOmmSIyrUv3ShLiCM%2F9MRpB)) na uvažovaný trojhuolník, čím dostaneme 

$$||\mathbf{u}-\mathbf{v}||^2=||\mathbf{u}||^2+||\mathbf{v}||^2-2||\mathbf{u}||||\mathbf{v}||\cos\theta.$$ 

Rozpísaním ľavej strany pomocou vzťahu $||\mathbf{u}-\mathbf{v}||^2=(\mathbf{u}-\mathbf{v})\cdot(\mathbf{u}-\mathbf{v})$ dostaneme 

$$||\mathbf{u}||^2-2(\mathbf{u}\cdot\mathbf{v})+||\mathbf{v}||^2=||\mathbf{u}||^2+||\mathbf{v}||^2-2||\mathbf{u}||||\mathbf{v}||\cos\theta.$$ 

Úpravou poslednej rovnosti prídeme k požadovanému vzťahu pre kosínus odchýlky dvoch vektorov $\mathbf{u}$, $\mathbf{v}$ 

$$\cos\theta=\dfrac{\mathbf{u}^T\cdot\mathbf{v}}{||\mathbf{u}||\cdot||\mathbf{v}||}.$$ 

Naznačili sme vlastne dôkaz Tvrdenia 2. 

---

#### <font color=green> Tvrdenie 2 </font>

Pre nenulové vektory $\mathbf{u},\mathbf{v}\in\mathbb{R}^n$ platí 

$$\cos\theta=\dfrac{\mathbf{u}^T\cdot\mathbf{v}}{||\mathbf{u}||\cdot||\mathbf{v}||}.$$

---

**Odchýlka vektorov vs uhol:** Na nasledujúcom obrázku vidíme dva uhly (jeden väčší a jeden menší), ktoré zvierajú vektory $\mathbf{u}$, $\mathbf{v}$. Odchýlka je menší z nich. Presnejšie, odchýlka vektorov je uhol medzi vektormi, ktorý je z intervalu $\langle0,180^\circ \rangle$. Niekedy sa zvykne použiť výraz "uhol" ale myslí sa tým odchýlka. 

 


<font color=orange> Určte odchýlku (nie kosínus odchýlky) vektorov $\mathbf{u}=(2,1,-2)^T$, $\mathbf{v}=(1,1,1)$. </font>

Iste si vieme predstaviť **kolmé vektory** v 2D alebo v 3D. Kvôli jasnejšej predstave si môžeme pozrieť nasledujúce dva obrázky, kde v rovine ($\mathbb{R}^2$) máme znázornené dva kolmé vektory $\mathbf{u}=(2,5)^T$, $\mathbf{v}=(-5,2)^T$ a v trojrozmernom priestore ($\mathbb{R}^3$) máme znázornené dva kolmé vektory $\mathbf{u}=(1,2,-2)^T$, $\mathbf{v}=(-2,-1,-2)^T$. <br/> 
Zrejme v $\mathbb{R}^2$ alebo v $\mathbb{R}^3$ sú dva nenulové vektory $\mathbf{u}$, $\mathbf{v}$ kolmé, ak uhol $\theta$, ktorý zvierajú je $\theta=\dfrac{\pi}{2}$ alebo $90^\circ$. Preto vzhľadom k Definícii 2 máme $\dfrac{\mathbf{u}^T\cdot\mathbf{v}}{||\mathbf{u}||\cdot||\mathbf{v}||}=\cos90^\circ=0$, odkiaľ vyplýva, že musí byť $\mathbf{u}^T\cdot\mathbf{v}=0$, čo je vlastne motivácia k Definícii 2. <br/> 
<br/>

 # <font color=brown> Niekoľko geometrických interpretácii </font>

Teraz sa bližšie pozrieme na to, **čo môže znamenať súčin matice a vektora - ako to geometricky interpretovať**. Predpokladajme, že máme maticu $\mathbf{A}$, ktorou keď vynásobime vektor $\mathbf{x}$, tak dostaneme výsledný vektor $\mathbf{b}$. Formálne to môžeme zapísať takto $\mathbf{A}\mathbf{x}=\mathbf{b}$. Nasledujúce konkrétne príklady 

$$\begin{pmatrix}
4 & -3 & 1 & 3 \\
2 & 0 & 5 & 1
\end{pmatrix}
\begin{pmatrix}
1 \\
1 \\
1 \\
1
\end{pmatrix}=
\begin{pmatrix}
5 \\ 
8
\end{pmatrix}, \quad
\begin{pmatrix}
4 & -3 & 1 & 3 \\
2 & 0 & 5 & 1
\end{pmatrix}
\begin{pmatrix}
1 \\
4 \\
-1 \\
3
\end{pmatrix}=
\begin{pmatrix}
0 \\ 
0
\end{pmatrix},$$

nám vlastne hovoria, že vektor $\mathbf{x}=
\begin{pmatrix}
1 \\
1 \\ 
1 \\
1
\end{pmatrix}$ vynásobený maticou $
\mathbf{A}=\begin{pmatrix}
4 & -3 & 1 & 3 \\
2 & 0 & 5 & 1
\end{pmatrix}$ sa zmení (transformuje) na vektor $\mathbf{b}=
\begin{pmatrix}
5 \\
8
\end{pmatrix}$, resp.,  
že vektor $\mathbf{u}=
\begin{pmatrix}
1 \\
4 \\ 
-1 \\
3
\end{pmatrix}$ vynásobený maticou $
\mathbf{A}=\begin{pmatrix}
4 & -3 & 1 & 3 \\
2 & 0 & 5 & 1
\end{pmatrix}$ sa zmení (transformuje) na vektor $\mathbf{0}=
\begin{pmatrix}
0 \\
0
\end{pmatrix}$. Vo všeobecnosti teda hovoríme, že vektor $\mathbf{x}$ sa po vynásobení maticou $\mathbf{A}$ zobrazí na vektor $\mathbf{b}$ tak, ako to ilustruje nasledujúci obrázok. 


$$\text{Zdroj: Lay (2016).}$$

Ináč povedané, ak vezmeme z nejakej množiny vektorov konkrétny vektor $\mathbf{x}$, tak jemu je vďaka násobeniu maticou $\mathbf{A}$ priradený konkrétny vektor $\mathbf{A}\mathbf{x}$ z nejakej množiny vektorov. Čiže množinu dvojíc $\left\{[\mathbf{x},\mathbf{A}\mathbf{x}]\right\}$ resp. priradenie medzi $\mathbf{x}$ a $\mathbf{A}\mathbf{x}$ je možné chápať ako zobrazenie z jednej množiny vektorov do druhej množiny vektorov. To zobrazenie (jeho vlastnosti) je určené práve maticou $\mathbf{A}$. 

V našom konkrétnom príklade uvedenom vyššie sa štvorrozmerný vektor $\mathbf{x}$ zobrazí na dvojrozmerný vektor $\mathbf{A}\mathbf{x}=\mathbf{b}$. 

Ak sa nejaký vektor zobrazí na vektor rovnakej dimenzie (na vektor s rovnakým počtom zložiek), tak toto zobrazenie sa zvykne nazývať **transformácia**. Napr. keď sa vektor v rovine (v 2D) po vynásobení maticou (typu $2\times2$) zobrazí na nejaký (iný) vektor opať v rovine (v 2D).

\begin{pmatrix}

To znamená, že keď vezmeme do úvahy sčítanie vektorov a tiež násobenie vektorov konštantou, tak môžeme násobenie vektora maticou zapísať ako súčet troch vektorov - stĺpcov matice, ktoré sú vynásobené práve zložkami vektora. Konkrétne ako v našom príklade to je prvá zložka vektora krát prvý stĺpec matice plus druhá zložka vektora krát druý stĺpec matice plus tretia zložka vektora krát tretí stĺpec matice. Ináč povedané, výsledný vektor $(2,-2,-1)^{T}$ je **lineárnou kombináciou** stĺpcov matice, pričom koeficienty tejto lineárnej kombinacie sú práve zložky vektora $(3,2,-1)^{T}$, ktorý je maticou násobený. 

Tento princíp sa uplatňuje aj pri vektoroch a maticiach v 2D alebo aj pri viacrozmerných vektoroch a maticiach pre $n>3$ (kde **$n$ je počet zložiek vektora - teda jeho dimenzia**). To nás vedie k nasledujúcej definícii.

### <font color=red> Definícia 3 </font>

Vektor $\mathbf{v}\in\mathbb{R}^{n}$ je **_lineárnou kombináciou_** vektorov $\mathbf{v}_1,\ldots,\mathbf{v}_n\in\mathbb{R}^{n}$ ak existujú skaláre (konštanty) $p_1,\ldots,p_n\in\mathbb{R}$  také, že platí 

$$\mathbf{v}=p_1\mathbf{v}_1+\ldots+p_n\mathbf{v}_n.$$

Skaláre (čísla) $p_1,\ldots,p_n$ sa nazývajú **_koeficienty_** lineárnej kombinácie.

#### Poznámka 7

* Vektor, ktorý je výsledkom súčtu ľubovoľného konečného počtu vektorov je vlastne tiež lineárnou kombináciou týchto vektorov. Koeficienty takej lineárnej kombinácie sú všetky rovné $1$. Napr. 

$$\left(\begin{array}{r}
1 \\
0 \\
0
\end{array}\right)+
\left(\begin{array}{r}
0 \\
1 \\
0
\end{array}\right)+
\left(\begin{array}{r}
0 \\
0 \\
1
\end{array}\right)=
1\cdot\left(\begin{array}{r}
1 \\
0 \\
0
\end{array}\right)+
1\cdot\left(\begin{array}{r}
0 \\
1 \\
0
\end{array}\right)+
1\cdot\left(\begin{array}{r}
0 \\
0 \\
1
\end{array}\right)=
\left(\begin{array}{r}1 \\
1 \\
1
\end{array}\right)$$


* Ak existujú koeficienty (reálne čísla) $p_1,\ldots,p_n\in\mathbb{R}$ také, že vektor $\mathbf{v}\in\mathbb{R}^n$ je možné vyjadriť ako lineárnu kombináciu vektorov $\mathbf{v}_1,\ldots\mathbf{v}_n\in\mathbb{R}^n$ t.j. v tvare $\mathbf{v}=p_1\mathbf{v}_1+\ldots+p_n\mathbf{v}_n$, tak hovoríme, že **vektory $\mathbf{v}_1,\ldots\mathbf{v}_n,\mathbf{v}$ sú lineárne závislé**. V opačnom prípade sú tieto **vektory lineárne nezávislé**. Inými slovami, ak máme konečný počet vektorov a jeden z nich je možné vyjadriť ako lineárnu kombináciu ostatných vektorov, tak všetky tieto vektory spolu tvoria množinu lineárne závislých vektorov. 

---

Ekvivalentnú (i keď na pohľad inú) charakterizáciu lineárne (ne)závislých vektorov ponúka nasledujúca definícia. Ak v Poznámke 7 (druhá časť) premiestnime vektor $\mathbf{v}$ na jednú stranu k vektorom $\mathbf{v}_1,\ldots\mathbf{v}_n$, dostaneme tak vyjadrenie $p_1\mathbf{v}_1+\ldots+p_n\mathbf{v}_n-\mathbf{v}=\mathbf{0}$. Tzn. **nulový vektor $\mathbf{0}$ je vyjadrený ako netriviálna lineárna kombinácia vektorov $\boldsymbol{v}_1,\ldots\boldsymbol{v}_n,\boldsymbol{v}$**. Netriviálna v zmysle, že aspoň jeden z koeficientov lineárnej kombinácie je nenulový, v tomto prípade je to určite aspoň koeficient $-1$ pri vektore $\mathbf{v}$. V takom prípade **vektory $\mathbf{v}_1,\ldots\mathbf{v}_n,\mathbf{v}$** považujeme za **lineárne závislé**. V opačnom prípade, teda ak by **nulový vektor $\boldsymbol{0}$** bolo **možné vyjadriť len ako triviálnu lineárnu kombináciu** (všetky koeficienty lineárnej kombinácie nulové) pomocou vektorov $\mathbf{v}_1,\ldots\mathbf{v}_n,\mathbf{v}$ t.j. $0\cdot\mathbf{v}_1+\ldots+0\cdot\mathbf{v}_n+0\cdot\mathbf{v}=\mathbf{0}$, v takom prípade nazveme **vektory $\mathbf{v}_1,\ldots\mathbf{v}_n,\mathbf{v}$ lineárne nezávislé**. V Definícii 5 sa uvažujú len vektory $v_1,\ldots,v_n$ ale princíp je rovnaký. Podstatné je, že máme **vždy iba konečný počet vektorov** (nie je podstatné koľko ich presne je). 

---

### <font color=red> Definícia 4 </font>

Vektory $\mathbf{v}_1,\ldots,\mathbf{v}_n\in\mathbb{R}^{n}$ nazývame **_lineárne závislé_**, ak existujú skaláre (konštanty) $p_1,\ldots,p_n\in\mathbb{R}$  také, že aspoň jedno $p_i\neq0$ (pre nejaké $i\in\{1,\ldots n\}$) a súčasne platí 

$$p_1\mathbf{v}_1+\ldots+p_n\mathbf{v}_n=\mathbf{0}.$$

V opačnom prípade nazývame vektory $\mathbf{v}_1,\ldots,\mathbf{v}_n$ **_lineárne nezávislé_** t.j. v prípade, ak  existuje iba jediné vyjadrenie 

$$p_1\mathbf{v}_1+\ldots+p_n\mathbf{v}_n=\mathbf{0}$$

a to také, kde $p_1=p_2=\ldots=p_n=0$.

#### Poznámka 8

Pre **overenie lineárnej (ne)závislosti množiny daných vektorov** $\mathbf{v}_1,\ldots,\mathbf{v}_n\in\mathbb{R}^{n}$, stačí si vzhľadom k Definícii 4 prepísať podmienku $p_1\mathbf{v}_1+\ldots+p_n\mathbf{v}_n=\mathbf{0}$ v maticovom tvare $\mathbf{V}\mathbf{p}=\mathbf{0}$, kde $\mathbf{V}=(\mathbf{v}_1\, \mathbf{v}_2\,\ldots\,\mathbf{v}_n)$ t.j. vektory uložené po stĺpcoch vedľa seba do matice, $\mathbf{p}=(p_1,\ldots,p_n)^T$, $\mathbf{0}$ je nulový vektor a vyriešiť túto homogénnu sústavu lineárnych rovníc (t.j. s nulovou pravou stranou), kde neznáme sú koeficienty  $p_1,\ldots,p_n\in\mathbb{R}$. Podľa počtu a povahy  riešení uvedenej sústavy rovníc sa urobí vzhľadom k Definícii 4 záver o lineárnej (ne)závislosti vektorov $\mathbf{v}_1,\ldots,\mathbf{v}_n$. Sústavu lineárnych rovníc môžeme zatiaľ riešiť s pomocou Sage.

#### <font color=blue> Príklad 3 </font>

Zistite, či sú vektory $\mathbf{v}_1,\mathbf{v}_2,\mathbf{v}_3\in\mathbb{R}^4$ lineárne závislé, ak sú tieto vektory dané nasledovne: $\mathbf{v}_1=(-2,2,0,1)^T$, $\mathbf{v}_2=(-4,4,1,2)^T$, $\mathbf{v}_3=(2,-2,-1,1)^T$. 

<ins> Riešenie: </ins> 

Podľa vyššie uvedenej teórie máme zistiť, či existujú konštanty (koeficienty) $p_1,p_2,p_3$ - nie všetky rovné nule tak, aby platilo 

$$p_1\mathbf{v}_1+p_2\mathbf{v}_2+p_3\mathbf{v}_3=\mathbf{0},$$ 

pričom po dosadení vektorov máme tvar 

$$p_1\cdot\begin{pmatrix}-2\\2\\0\\1\end{pmatrix}+p_2\cdot\begin{pmatrix}-4\\4\\1\\2\end{pmatrix}+p_3\cdot\begin{pmatrix}2\\-2\\-1\\1\end{pmatrix}=\begin{pmatrix}0\\0\\0\\0\end{pmatrix}.$$

Ak vektory vynásobime koeficientmi ($p_1,p_2,p_3$) a sčítame, tak dostaneme tvar 

$$\begin{pmatrix}-2p_1-4p_2+2p_3\\2p_1+4p_2-2p_3\\0p_1+1p_2-1p_3\\1p_1+2p_2+1p_3\end{pmatrix}=\begin{pmatrix}0\\0\\0\\0\end{pmatrix},$$

pričom teraz máme na pravej i na ľavej strane jeden vektor a medzi nimi znak "$=$". Dva vektory sa rovnajú práve vtedy, keď sa rovnajú ich odpovedajúce zložky a teda rozpísaním po zložkách dostávame sústavu štyroch lineárnych rovníc s troma neznámymi ($p_1,p_2,p_3$) v tvare 

\begin{align}
-2p_1-4p_2+2p_3 & = 0 \\ 
2p_1+4p_2-2p_3 & = 0 \\ 
p_2-p_3 & = 0 \\ 
p_1+2p_2+p_3 & = 0.  
\end{align}

K tej istej sústave rovníc by sme sa dopracovali, keby sme si zapísali sústavu rovníc v maticovom tvare podľa Poznámky 8 a to nasledovným spôsobom: 

$$\mathbf{V}\mathbf{p}=\mathbf{0},$$ 

čo je v našom konkrétnom príklade tento tvar 

$$\begin{pmatrix}-2&-4&2\\2&4&-2\\0&1&-1\\1&2&1\end{pmatrix}\cdot\begin{pmatrix}p_1\\p_2\\p_3\end{pmatrix}=\begin{pmatrix}0\\0\\0\\0\end{pmatrix},$$

ďalej vynásobením matice a vektora dostaneme rovnaký tvar ako je uvedený vyššie t.j. 

$$\begin{pmatrix}-2p_1-4p_2+2p_3\\2p_1+4p_2-2p_3\\0p_1+1p_2-1p_3\\1p_1+2p_2+1p_3\end{pmatrix}=\begin{pmatrix}0\\0\\0\\0\end{pmatrix}.$$

```python
# riesenie sustavy linearnych rovnic pomocou Sage v sulade s Poznamkou 8
# zavedenie premennych - neznamych koeficientov linearnej kombinacie p1, p2, p3
p1 = var('p1', latex_name='p_1') 
show(p1) 
p2 = var('p2', latex_name='p_2') 
show(p2) 
p3 = var('p3', latex_name='p_3') 
show(p3) 
p = vector([p1,p2,p3]) # vektor p 
show(p) 
V = matrix(4,3,[-2,-4,2,2,4,-2,0,1,-1,1,2,1]) # matica V - stlpce su vektory v1, v2, v3
show(V) 
o = vector([0,0,0,0]) # nulovy vektor - prava strana 
show(o)
eq = [[(V*p)[k] == o[k]] for k in [0,1,2,3]] # vytvorenie sustavy rovnic 
# iny sposob ako zadat sustavu rovnic v Sage 
# eq = [[-2*p1-4*p2+2*p3 == 0], [2*p1+4*p2-2*p3 == 0], [0*p1+1*p2-1*p3 == 0], [1*p1+2*p2+1*p3 == 0]]
show(eq) 
eq_sol = solve(eq, p1, p2, p3) # vyriesenie sustavy rovnic - najdenie koeficientov p1, p2, p3 
show(eq_sol)
```

Jediné riešenie uvedenej sústavy lineárnych rovníc je $p_1=0$, $p_2=0$, $p_3=0$. Podľa Definície 4 resp. podľa poznámky pred ňou to znamená, že zadané vektory $\mathbf{v}_1$, $\mathbf{v}_2$, $\mathbf{v}_3$ sú lineárne nezávislé, keďže nulový vektor z nich môžeme dostať len ako triviálnu lineárnu kombináciu t.j. 

$$0\cdot\mathbf{v}_1+0\cdot\mathbf{v}_2+0\cdot\mathbf{v}_3=\mathbf{0}.$$

#### <font color=green> Tvrdenie 4 </font>

Nech $\mathbf{a},\mathbf{b},\mathbf{c}\in\mathbb{R}^3$, ďalej nech $\theta$ je odchýlka vektorov $\mathbf{a}$, $\mathbf{b}$ a $p$ nech je reálna konštanta. Potom platia nasledujúce vzťahy: 

(i) $||\mathbf{a}\times\mathbf{b}||=||\mathbf{a}||\cdot||\mathbf{b}||\cdot\sin\theta$ (obsah rovnobežníka so stranami $\mathbf{a}$, $\mathbf{b}$); 


(ii) $\mathbf{a}\times\mathbf{b}=-\mathbf{b}\times\mathbf{a}$; 


(iii) $(p\mathbf{a})\times\mathbf{b}=p(\mathbf{a}\times\mathbf{b})=\mathbf{a}\times(p\mathbf{b})$; 


(iv) $\mathbf{a}\times(\mathbf{b}+\mathbf{c})=\mathbf{a}\times\mathbf{b}+\mathbf{a}\times\mathbf{c}$; 


(v) $\mathbf{a}^T\cdot(\mathbf{b}\times\mathbf{c})=(\mathbf{a}\times\mathbf{b})^T\cdot\mathbf{c}$.

Majme rovnobežník určený vektormi $\mathbf{a}$, $\mathbf{b}$, pričom uhol medzi nimi je $\theta$. Zrejme **obsah rovnobežníka** vypočítame ako dĺžka strany krát dĺžka výšky na príslušnú stranu. Vezmime stranu rovnobežníka s dĺžkou $||\mathbf{a}||$ a výšku $v$ na túto stranu, ktorú vďaka definícii sínusu uhla v pravouhlom trojuholníku určíme zo vzťahu $\sin\theta=\dfrac{v}{||\mathbf{b}||}$, odkiaľ $v=||\mathbf{b}||\cdot\sin\theta$. **Obsah rovnobežníka** je teda "strana krát výška" t.j. 

$$||\mathbf{a}||\cdot v=||\mathbf{a}||\cdot||\mathbf{b}||\sin\theta$$ 

**a to je podľa Tvrdenia 3 časti (i) rovné výrazu $||\mathbf{a}\times\mathbf{b}||$, čím dostávame jednoduchú geometrickú  interpretáciu vektorového súčinu resp. jeho normy - ako obsah rovnobežníka so stranami $\mathbf{a}$, $\mathbf{b}$**. <br/> 
Presne túto situáciu znázorňujú obrázky vyššie, akurát na prvom z nich sú vektory označené aj so šípkou ($\overset{\rightarrow}{\mathbf{a}}$, $\overset{\rightarrow}{\mathbf{b}}$) a na druhom je norma vektora označená ako absolútna hodnota (tzn. po jednej zvislej čiare na oboch stranách miesto dvoch, t.j. $|\mathbf{a}\times\mathbf{b}|$ miesto $||\mathbf{a}\times\mathbf{b}||$). <br/> 
Existujú i ďalšie zaujímavé interpretácie vektorového súčinu (napr. taká, ktorá dáva do súvisu vektorový súčin so skalárnym súčinom), no tie sú už nad rámec nášho kurzu.
