Ktorý z týchto tímov je lepší? Ako by sme mali porovnať dva tímy, ktoré sa proti sebe nikdy nestretnú? Mali by sa zohľadnovať nazbierané body resp. skóre zo zápasov?  Napriek týmto zložitostiam sa poradie športovcov a športových tímov stalo samozrejmosťou a veľmi očakávanou zložkou v médiách. Práve špeciálne **vektory** - tzv. **vlastné vektory** (podrobnejšie v téme 7) a teda aj určité špeciálne **matice** (i tzv. modely z teórie grafov) hrajú dôležitú úlohu **v algoritmoch** riešiacich naznačené otázky.

Ďalšou oblasťou, kde sa uplatňuje **analytická geometria** je **vytváranie nových obrázkov zo starých obrázkov**. Majme pre jednoduchosť dva čiernobiele obrázky, muža a ženu reprezentované ako postupnosti čísel teda vektory $\mathbf{u}$, $\mathbf{v}$.  

 
$$\text{Zdroj: Klein P.N. (2013).}$$

Nový obrázok môžeme vytvoriť napr. ako **lineárnu kombináciu** pôvodných dvoch obrázkov resp. **vektorov $\mathbf{u}$, $\mathbf{v}$** napr. v tvare $\frac{1}{2}\mathbf{u}+\frac{1}{2}\mathbf{v}$. K tomu ale potrebujeme rozumieť **lineárnym kombináciám vektorov** a vôbec **operáciám s vektormi**. Výsledný produkt je znázornený na nasledujúcom obrázku. Je to akýsi "priemer z dvoch obrázkov".  


$$\text{Zdroj: Klein P.N. (2013).}$$ 

Začneme najprv **geometrickým pohľadom na vektory a operácie** s nimi. Následne si ukážeme ako určiť **normu** (veľkosť resp. dĺžku) **vektora** a tiež **odchýlku** dvoch **vektorov**. Pozrieme sa aj na to, čo môže **geometricky** vyjadrovať **súčin matice s vektorom** a čo reprezentuje **determinant**. Tieto geometrické predstavy a interpretácie nám ďalej pomôžu pri lepšom pochopení **aplikácií lineárnej algebry v informatike** (v ďalšej téme). 

Budeme sa zaoberať aj inými dôležitými pojmami z hľadiska aplikácií lineárnej algebry a geometrie v informatike, konkrétne pôjde o **lineárne kombinácie vektorov**, **lineárnu (ne)závislosť vektorov** či **skalárny** i **vektorový súčin vektorov**. Tieto pojmy budeme ilustrovať na príkladoch a objasníme aj ich geometrickú interpretáciu.
