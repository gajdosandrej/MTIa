# <font color=brown> Motivácia</font>

Špeciálne postavenie v **teórii čísel** ale tiež v **šifrovaní** majú **prvočísla**. V rámci témy 2 nadviažeme na deliteľnosť a objasníme jej ďalšie vzťahy súvisiace s prvočíslami. Priblížime si **testovanie prvočíselnosti** resp. **hľadanie rozkladu zloženého čisla na prvočísla**, k čomu dáva základ tzv **Základná veta aritmetiky**. Budeme ďalej budovať matematický aparát potrebný pre aplikácie v spomínanom [šifrovaní](https://cloud2q.edupage.org/cloud/Sifrovanie.pdf?z%3AQVD4sH%2Bxg8wspsXbJ60dIiO904pzfdtdIkOHTLtRR6pBh0KvbJs87LWhWM8E1O5Y) či [kódovaní](http://www.spslevice.sk/ucebnice/soc-uceb-inf2007/ucebnica_inf/informatika/informacia/informacia.htm), a preto sa oboznámime aj s konceptom podobným rovniciam - tzv. **kongruenciami**. Ukážeme si ako je možné využiť **kongruencie pri uľahčení náročných výpočtov** dokonca aj tam, kde už výpočtová sila počítačov môže zlyhávať.
