# <font color=brown> Motivácia </font>

Predstavme si, že máme vyčísliť / **odhadnúť čas výpočtu** nejakého **programu** resp. **časovú** (výpočtovú) **zložitosť** nejakého **algoritmu** určeného pre riešenie istého typu praktickej úlohy. Môže sa jednať o **triediaci algoritmus** (bubble sort, quick sort, heap sort, merge sort, ... ) t.j. zoradenie prvkov od najmenšieho po najväčší (a vice versa) alebo abecedne (lexikograficky), prípadne môže ísť o **vyhľadávací algoritmus** (nájdenie zadanej hodnoty v zozname resp. poli - binárne vyhľadávanie, lineárne vyhľadávanie, ...), či **grafový algoritmus** (prehľadávanie do šírky, prehľadávanie do hĺbky, hľadanie najkratšej cesty, najlacnejšej kostry, ...), kde graf chápeme ako množinu vrcholov (bodiek resp. guličiek), ktoré sú pospájané hranami (čiarami) - viď obrázok nižšie. 
Vo všetkých uvedených prípadoch a v mnohých ďalších, je **časová zložitosť algoritmu funkciou** veľkosti vstupu $n$ (tzn. časová zložitosť resp. výpočtový čas závisí od $n$), pričom $n$ môže reprezentovať veľkosť poľa resp. dĺžku zoznamu prvkov, počet vrcholov alebo hrán v grafe a pod. 

 
$$\text{Petersenov graf}$$

S **funkciami** je možné sa strenúť aj pri riešení **klasifikačných úloh pomocou algoritmov strojového učenia** (_machine learning_). Napríklad keď sa má banka rozhodnúť na základe určitých kritérii (charakteristík klienta), či mu poskytne alebo neposkytne pôžičku. Rovnako dôležité sú **funkcie** a ich kombinovanie (**skladanie**) aj v oblasti **umelej inteligencie** (s využitím napr. hĺbkového učenia pomocou neurónových sietí - tzv. _deep learning_), čo nachádza uplatnenie pri **vývoji a konštrukcii samojazdiacich áut**, **parkovacích kamier a senzorov**, **pri rozlišovaní dopravných značení v navigačných systémoch GPS**, či **v medicíne** pri určovaní typu alebo štádia ochorenia na základe snímkov. 

 

Nech už sa pozrieme kdekoľvek, všade číhajú nejaké **funkcie**. Ide o veľmi dôležitý koncept, preto je téma 9 venovaná práve **funkciám** a ich základným vlastnostiam ako napr. **párnosť**, **periodicita**, **monotónnosť**, **ohraničenosť**, ale aj ďalším dôležitým pojmom ako je **definičný obor**, **obor hodnôt funkcie** či **graf**. V neposlednom rade sa budeme v rámci tejto témy zaoberať i spôsobmi vytvárania nových funkcií z daných funkcií a to pomocou **operácií s funkciami**, za pomoci **skladania funkcií** či konceptu **inverznej funkcie**.
