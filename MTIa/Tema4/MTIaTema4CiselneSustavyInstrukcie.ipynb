<a id=motivacia></a>
# <font color=brown> Motivácia</font>

Pre nás ľudí je najprirodzenejšia tzv. **desiatková číselná sústava**. Stretávame sa a učíme sa s ňou narábať prakticky od škôlky. Ľudia, ktorí inklinujú k **informačným technológiám** a potrebujú porozumieť napr. princípom počítačov či iných zariadení by si mali osvojiť aj iné **číselné sústavy** ako napr. **dvojková**, **osmičková** alebo **šestnástková číselná sústava**. Totiž **digitálne zariadenia komunikujú v iných číselných sústavách** - predovšetkým v **dvojkovej (binárnej)** zloženej z **núl** a **jednotiek**. Napr. ak vyťukáte niečo na klávesnici počítača, tak to, čo ste zadali, je ihneď prevedené do inej číselnej sústavy, aby tomu počítač "rozumel".  

 

S **číselnými sústavami** úzko súvisia aj používané **kódovacie schémy** ako napr. ASCII, ISCII či UNICODE, o ktorých sa viac dozviete na inom predmete :-) **Kódovanie a rôzne číselné sústavy** majú aj čo-to dočinenia s **IP adresami** (napr. v rámci [CIDR](https://www.keycdn.com/support/what-is-cidr) - [Classless Inter Domain Routing](https://www.techtarget.com/searchnetworking/definition/CIDR)) alebo s **reprezentáciou farbieb v počítači** (RGB - Red Green Blue). Číselné sústavy (konkrétne **hexadecimálna**) sa uplatňujú tiež pri procesoch akými sú napr. tzv. **adresovanie pamäte v PC** alebo **šifrovanie komunikácie** z hľadiska bezpečnosti. 

V rámci témy 4 sa oboznámime so základnými **pozičnými číselnými sústavami** používanymi v informačných technológiách - **desiatková (decimálna), dvojková (binárna), osmičková (oktálna), šestnástková (hexadecimálna) číselná sústava**. Ukážeme si ako **reprezentovať čísla** v týchto sústavách a ako robiť **prevody medzi jdnotlivými reprezentáciami čísel v rôznych sústavách**. V závere si ešte na príkladoch objasníme základné myšlienky pre **artiemtické operácie** v číselných sústavách.
