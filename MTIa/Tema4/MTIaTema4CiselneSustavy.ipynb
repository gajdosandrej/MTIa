 # <font color=brown> Motivačný úvod </font> 

**Údaje** sú v technických zariadeniach uchovávané **ako postupnosti núl a jednotiek**. Takýto spôsob zápisu, ktorý používa na zápis informácií nuly a jednotky sa nazýva **binárny kód**. **Jednotkou informácie** je tzv. jeden **bit** (binary digit), ktorý má hodnotu buď $0$ lebo $1$. Nie je totiž možné komunikovať s menej ako dvoma stavmi, tzn. $0$/$1$ resp. áno/nie resp. vypnutý/zapnutý resp. true/false a pod.  

Zamysleli ste sa niekedy nad tým, **ako počítačový systém sám pre seba interpretuje klávesy na klávesnici počítača**, ktoré sú v rozpoznateľnej forme pre človeka? Táto časť stručne pojednáva o interpretácii textu počítačom. Možno viete, že počítač "rozumie" iba binárnemu jazyku t.j. "chápe" len nuly a jednotky. Preto, keď stisnete klávesu na klávesnici, táto klávesa má priradený identifikátor (určitý kód), ktorý sa následne prevedie do binárnej formy pozostávajúcej zo samých núl a jednotiek. V konečnom dôlsedku je teda konkrétna klávesa stlačená na klávesnici zakódovaná (konečnou) postupnosťou núl a jednotiek. Inými slovami, každá klávesa na klávesnici sa zobrazí do nejakej binárnej postupnosti, resp. každej klávese je jednoznačne priradená určitá binárna postupnosť. Tento proces názorne ilustruje Príklad 1.

**Čo je to teda kódovanie? Mechanizmus prevodu údajov na ekvivalentnú šifru pomocou špecifického kódu sa nazýva kódovanie.** 
Je dôležité pochopiť, prečo sa pre klávesu "A" používa kódová hodnota $65$ a nie iná hodnota? Je to rovnaké pre všetky klávesnice bez ohľadu na ich značku? Áno, je to rovnaké pre všetky klávesnice. To bolo možné vďaka štandardným schémam kódovania, kde je každé písmeno, číslica a symbol kódované resp. im je priradený jedinečný kód. Existuje viacero známych a používaných schém kódovania. My spomenieme jednu z nich - tzv. **ASCII** (American Standard Code for Information Interchange). Ďalšie schémy kódovania sú napr. [ISCII](https://en.wikipedia.org/wiki/Indian_Script_Code_for_Information_Interchange) (Indian Script Code for Information Interchange) alebo [UNICODE](https://en.wikipedia.org/wiki/Unicode). 

Na začiatku šesťdesiatych rokov minulého storočia počítače medzi sebou nemohli komunikovať kvôli rôznym spôsobom zobrazovania klávesov na klávesnici. Preto sa realizovala potreba spoločného štandardu na prekonanie tohto nedostatku. Schéma kódovania ASCII bola teda vyvinutá pre štandardizáciu reprezentácie znakov. ASCII je stále najbežnejšie používanou kódovacou schémou. 

Pôvodne ASCII používalo na reprezentáciu znakov 7 bitov. Pripomeňme, že existujú iba 2 binárne číslice ($0$ alebo $1$). Celkový počet rôznych znakov na anglickej klávesnici, ktoré je možné kódovať 7-bitovým kódom ASCII, je teda $2^7 = 128$. Nasledujúca  tabuľka uvádza niektoré tlačiteľné znaky pre kód ASCII. ASCII však môže kódovať iba znakovú sadu v angličtine. Stĺpce v tabulke s názvom "Character" obsahujú rôzne znaky (z klávesnice) a stĺpce s názvom "Decimal Value" obsahujú ASCII kód pre príslušný znak.

 # <font color=brown> Číselné sústavy</font> 

Zatiaľ sme sa dozvedeli, že každý kláves (predstavujúci znak, špeciálny symbol, funkčné klávesy atď.) klávesnice je interne mapovaný (zobrazený) na kód ASCII podľa schémy kódovania. Táto zakódovaná hodnota sa ďalej prevedie na ekvivalentnú binárnu reprezentáciu, aby ju počítač pochopil. Na obrázku z úvodu témy kód $65$ pre znak "A" patrí do tzv. desiatkovej sústavy (s týmito číslami my ľuda bežne pracujeme) a jeho ekvivalentná binárna hodnota patrí do systému binárnych čísel t.j. do tzv. dvojkovej resp. binárnej sústavy. **Číselná sústava** je spôsob reprezentácie (zapisovania) čísel. 

**Každá číselná sústava má množinu jedinečných znakov alebo literálov.** Počet týchto literálov sa nazýva **_radix_** alebo **_základ číselnej sústavy_**. Štyri rôzne číselné sústavy používané v kontexte počítačov sú znázornené na ďalšom obrázku. Konkrétne ide o **binárnu (dvojkovú) sústavu, oktálnu (osmičkovú) sústavu, decimálnu (desiatkovú) sústavu a hexadecimálnu (šesnástkovú) sústavu**.  Tieto číselné sústavy si postupne popíšeme v nasledujúcich častiach učebného textu. 

 
$$\text{Zdroj: NCERT (2019).}$$ 

Číselné sústavy, korým sa budeme venovať v rámci nášho kurzu, sa tiež nazývajú **pozičné číselné sústavy**, pretože hodnota každého symbolu (t. j. číslice alebo písmena abecedy) v čísle závisí od jeho polohy (pozície) v rámci čísla. Číslo môže mať aj zlomkovú časť podobnú desatinným číslam, ktoré používame. Symbol úplne vpravo v celočíselnej časti v danom čísle má pozíciu 0. Hodnota polohy resp. pozície v celočíselnej časti sa zvyšuje sprava doľava o 1. Na druhej strane prvý symbol v zlomkovej časti čísla (napr. za desatinnou čiarkou vpravo) má pozíciu -1, ktorá sa pri čítaní zlomkovej časti zľava doprava zníži o 1. Každý symbol v čísle má **pozičnú hodnotu**, ktorá sa vypočíta pomocou hodnoty jeho polohy a pomocou tzv. základu uvažovanej číselnej sústavy. Symbol na pozícii číslo 3 v desiatkovej sústave so základom $10$ má pozičnú hodnotu $10^3$. **Sčítaním súčinu pozičnej hodnoty a hodnoty symbolu vznikne dané číslo.** Predchádzajúce úvahy sú ilustrované v nasledujúcej tabuľke, ktorá ukazuje výpočet desatinného čísla $123.45$ pomocou jeho pozičnej hodnoty (v štandardnej desiatkovej sústave, kde tzv. základom je číslo $10$). 

 
$$\text{Zdroj: NCERT (2019).}$$ 

Na základe tabuľky (zľava doprava) a predošleho odseku dostávame zápis čísla $123.45$ v desiatkovej sústave pomocou mocnín základu - t.j. mocnín desiatky (resp. pomocou súčtu súčinov pozičných hodnôt a cifier - číslic) t.j. $ 1\times10^2+2\times10^1+3\times10^0+4\times10^{-1}+5\times10^{-2}=(123.45)_{10}$ , čo je možné skrátiť a zapísať v tvare $123.45$. 

## Desiatková sústava

Desiatková (decimálna) sústava sa používa v našom každodennom živote. Je známa ako sústava so základom $b=10$ ($b$ z anglického _base_), pretože sa v nej používa $10$  číslic ($0$ až $9$). Číslo je reprezentované dvoma hodnotami - hodnotou symbolu (ľubovoľná číslica od $0$ do $9$) a pozičnou hodnotou (v zmysle základu $10$). Pozičná hodnota je vyjadrená v mocninách čísla $10$. Ďalší obrázok ukazuje celočíselnú a zlomkovú časť desatinného čísla $237.25$ spolu s výpočtom  pomocou pozičných hodnôt. 

 
$$\text{Zdroj: NCERT (2019).}$$ 

Ak to zhrnieme, tak dostávame zápis čísla $237.25$ pomocou mocnín základu - desiatky (resp. pomocou súčtu súčinov pozičných hodnôt a číslic) t.j.  $2\times10^2+3\times10^1+7\times10^0+2\times10^{-1}+5\times10^{-2}=(237.25)_{10}$. 

<font color=orange>  Podľa príkladu s číslom $123.45$ alebo s číslom $237.25$ zapíšte číslo $8204.11$ v desiatkovej sústave pomocou mocnín základu - t.j. pomocou mocnín čísla $10$ (resp. pomocou súčtu súčinov pozičných hodnôt a cifier - číslic). </font>

## Dvojková sústava 

ICs (integrované obvody) v počítači sú tvorené veľkým počtom tranzistorov, ktoré sú aktivované elektronickými signálmi (nízky/vysoký), ktoré prijímajú. Stav ON/ high a OFF/ low tranzistora je vyjadrený pomocou dvoch číslic $1$ a $0$. Tieto dve číslice $1$ a $0$ tvoria binárnu (dvojkovú) číselnú sústavu. Tento systém sa nazýva aj systém so základom $2$, pretože má iba dve číslice. Pozičná hodnota je vyjadrená v mocninách čísla $2$. Niektoré príklady binárnych čísel sú $(1001011)_2$, $(1011.101)_2$, $(111111.01)_2$. Binárnemu číslu je možné priradiť odpovedajúce číslo v desiatkovej sústave, ktoré je pre človeka ľahko pochopiteľné - viď nasledujúca tabuľka. 

 
$$\text{Zdroj: NCERT (2019).}$$  

Napr. $0=0\times2^0$ t.j. $(0)_2$; $1=1\times2^0$ t.j. $(1)_2$; $2=0\times2^0+1\times2^1$ t.j. $(10)_2$;  $4=0\times2^0+0\times2^1+1\times2^2$ t.j. $(100)_2$; $9=1\times2^0+0\times2^1+0\times2^2+1\times2^3$ t.j. $(1001)_2$. 

Ďalej napr. $(1011.101)_2$ je možné rozpísať pomocou mocnín základu - t.j. mocnín dvojky (resp. pomocou súčtu súčinov pozičných hodnôt a číslic) do tvaru $1\times2^3+0\times2^2+1\times2^1+1\times2^0+1\times2^{-1}+0\times2^{-2}+1\times2^{-3}=11.625$. Teda číslu $(1011.101)_2$ v dvojkovej sústave zodpovedá čislo $(11.625)_{10}$ v desiatkovej sústave. 

```python
# kontrola v Sage 
x = 11.625
show(x.str(base=2)) 
# vsimnime si, ze Sage pridal za desatinnu bodku 101 a potom vela nul, tieto nuly si v praxi mozeme odmysliet, 
# suvisi to s presnostou, s akou Sage pracuje 
```

```python
# kontrola v Sage 
y = 9
show(y.str(base=2))
```

## Osmičková sústava 

S nárastom hodnoty čísla v desiatkovej sústave sa zvyšuje aj počet bitov (núl a jednotiek) v jeho binárnom zápise - viď predošlá tabuľka. Binárne číslo je niekedy také veľké, že je ťažké s ním pracovať. Osmičková (oktálna) sústava čísel bola navrhnutá pre kompaktné znázornenie binárnych čísel. Systém osmičkových čísel sa nazýva sústava so základom $8$, pretože má celkom osem číslic ($0$-$7$) a pozičná hodnota je vyjadrená v mocninách čísla $8$. Tri binárne číslice (s doplnením núl v binárnom zápise) postačujú na reprezentáciu akejkoľvek číslice z osmičkovej sústavy (viď nasledujúca tabuľka), keďže $2^3=8$ (na každú z troch pozícii je možné dať buď $1$ alebo $0$). Ďalšia tabuľka zobrazuje desiatkové a binárne ekvivalenty číslic osmičkovej sústavy. Príklady čísel v osmičkovej sústave sú $(237.05)_8$, $(13)_8$ a $(617.24)_8$. 

 
$$\text{Zdroj: NCERT (2019).}$$  

Zápis čísla $(237.05)_8$ pomocou mocnín základu - t.j. mocnín čísla $8$ (resp. pomocou súčtu súčinov pozičných hodnôt a číslic) je $2\times8^2+3\times8^1+7\times8^0+0\times8^{-1}+5\times8^{-2}=159.078125$. Teda číslu $(237.05)_8$ v osmičkovej sústave zodpovedá číslo $(159.078125)_{10}$ v desiatkovej sústave. 

```python
# kontrola pomocou Sage 
z = 159.078125
show(z.str(base=8)) # opat nuly na konci je mozne si odmysliet, nemenia hodnotu cisla 
```

**Základ číselnej sústavy** sa používa na odlíšenie reprezentácie čísla v jednej číselnej sústave od reprezentácie čísla v inej  číselnej sústave. Základ sa zvykne zapisovať ako dolný index daného čísla. Napríklad $(70)_8$ predstavuje $70$ ako číslo reprezentované v osmičkovej sústave a $(70)_{10}$ označuje $70$ ako číslo v desiatkovej sústave. Asi tušíme, že ak by sme $(70)_8$ previedli z osmičkovej do desiatkovej sústavy, tak zrejme dostaneme v desiatkovej sústave niečo iné ako je číslo $70$. Ak je z kontextu jasné, o akú reprezentáciu čísla ide (v akej sústave), niekedy sa zvykne v zápise čísla základ (dolný index) vynechať. Zátvorky je tiež možné vynechať.

## Šestnástková sústava

Šestnástková (hexadecimálna) sústava sa taktiež používa aj na kompaktné zapisovanie binárnych čísel a prácu s nimi. Skladá sa zo 16 jedinečných symbolov ($0$-$9$, $A$ – $F$) a nazýva sa sústava so základom $16$. V šestnástkovej sústave je každá alfanumerická číslica (symbol) reprezentovaná ako skupina 4 binárnych číslic, pretože 4 bity ($2^4 = 16$) postačujú na reprezentáciu 16 alfanumerických symbolov. Všimnite si, že čísla 10 až 15 z desiatkovej sústavy sú reprezentované písmenami $A$ až $F$. Príklady hexadecimálnych čísel sú $(23A.05)_{16}$, $(1C3)_{16}$, $(619B.A)_{16}$. Nasledujúca tabuľka zobrazuje desiatkové a binárne  ekvivalenty 16 alfanumerických symbolov používaných v hexadecimálnej číselnej sústave. 

 
$$\text{Zdroj: NCERT (2019).}$$ 

Napr. číslo $(1C3)_{16}$ uvedené v šestnástkovej sústave, je možné zapísať pomocou mocnín základu - t.j. mocnín šestnástky (resp. pomocou súčtu súčinov pozičných hodnôt a číslic) nasledovne $1\times16^2+12\times16^1+3\times16^0=451$. Teda číslu $(1C3)_{16}$ v šestnástkovej sústave zodpovedá číslo $(451)_{10}$ v desiatkovej sústave. 

```python
# kontrola pomocou Sage 
x = 451
show(x.str(base=16)) 
```

Na záver tejto časti uvedieme dve **aplikácie hexadecimálnej čiselnej sústavy**. 

* Hlavná pamäť počítača sa skladá z pamäťových miest, kde každé miesto má jedinečnú adresu. Veľkosť pamäte je zvyčajne 16-bitová alebo 32-bitová (prípadne 64-bitová). Na prístup k adrese 16-bitovej pamäte musí programátor použiť 16 binárnych bitov, s čím je ťažké sa vysoriadať. Na zjednodušenie reprezentácie adresy sa používajú hexadecimálne a osmičkové čísla. Uvažujme 16-bitovú adresu pamäte $1100000011110001$. Pomocou hexadecimálnej notácie je táto adresa prevedená na $C0F1$, ktorá je ľahšie zapamätateľná. Ekvivalentná osmičková reprezentácia pre túto 16-bitovú hodnotu je $140361$. 


* Pre popis farieb na webovej stránke sa používajú aj hexadecimálne čísla. ### <font color=red> Definícia 1 </font>

Nech $b$ je prirodzené číslo väčšie ako $1$ a nech $n_0,n_1,\ldots,n_{b-1}$ je $b$ rôznych prirodzených čísel (nazývaných tiež číslice) takých, že $n_i$ označuje číslo $n_i=i$, pričom $i$ je z množiny $\{0,1,2,\ldots, b-1\}$. Reprezentácia prirodzeného čísla v číselnej sústave so základom $b$ je usporiadaná $n$-tica číslic $(d_nd_{n-1}\ldots d_{1}d_0)_d$, ktorá je interpretovaná ako prirodzené číslo nasledovným spôsobom 

$$d_nb^n+d_{n-1}b^{n-1}+d_{n-2}b^{n-2}+\ldots+d_1b^1+d_0b^0,$$

kde každá cifra $d_i$ vo vyjadrení čísla je rovná jednej z $b$ číslic $\{n_0,n_1,\ldots,n_{b-1}\}$, z ktorých číselná sústava so základom $b$ pozostáva. 

--- 

V príklade z Poznámky 3 máme **základ** $b=7$, ďalej číslice $n_0=0$, $n_1=1$, $n_2=2$, $\ldots$, $n_5=5$, $n_6=6$ a v prípade čísla $(13652)_7$ máme cifry $d_0=2$, $d_1=5$, $d_2=6$, $d_3=3$, $d_4=1$ i prislúchajúce **mocniny základu** (t.j. mocniny sedmičky) $b^0=7^0$, $b^1=7^1$, $b^2=7^2$, $b^3=7^3$, $b^4=7^4$. Potom **pozičná hodnota** v čísle $(13652)_7$ prislúchajúca cifre $2$ je $2\times7^0$, pozičná hodnota v čísle $(13652)_7$ prislúchajúca cifre $5$ je $5\times7^1$ atď. a nakoniec pozičná hodnota v čísle $(13652)_7$ prislúchajúca cifre $1$ je $1\times7^4$. **Súčtom** týchto **pozičných hodnôt** dostávame vyjadrenie čísla tak ako sme zvyknutí, t.j. v desiatkovej sústave v tvare $1\times7^4 +3\times7^3 +6\times7^2 +5\times7^1 +2\times7^0 = 3761$.

Desiatkovú sústavu čísel používajú ľudia najčastejšie, ale digitálne systémy rozumejú predovšetkým binárnym číslam, zatiaľ čo na zjednodušenie binárnej reprezentácie nám slúžia osmičkové a hexadecimálne číselné systémy. 


## Prevod z desiatkovej sústavy do inej číselnej sústavy 

**Pre konverziu čísla z desiatkovej sústavy do inej číselnej sústavy** so základom $b$ (napr. $b=2$ alebo $b=8$ alebo $b=16$) je možné **aplikovať nasledujúci postup** (veľmi pripomínajúci Euklidov algoritmus pre delenie celých čísel): 

**Krok 1:** Vydeliť dané číslo základom $b$ číselnej sústavy, do ktorej sa má previesť. 

**Krok 2:** Poznačiť zvyšok po delení. 

**Krok 3:** Pokračovať v delení kvocientu základom $b$, značiť si zvyšky po delení a skončiť, keď je kvocient rovný nule. 

**Krok 4:** Zapísať zvyšky po delení v opačnom poradí ako boli zaznačené t.j. zdola na hor a takto vznikne reprezentácia daného čísla v čiselnej sústave so základom $b$. 

#### Konverzia z desiatkovej do binárnej sústavy 

Pretože základom binárnej sústavy je číslo $2$, číslo z desiatkovej sústavy sa opakovane delí dvojkou podľa vyššie uvedených krokov, až kým kvocient nie je $0$. Je potrebné zaznamenať zvyšky po každom delení a nakoniec zapísať zvyšky v opačnom poradí, v akom sú vypočítané. Na začiatku v motivačnom príklade s počítačom a klávesnicou sme videli, že binárny ekvivalent čísla $65$ je $(1000001)_2$. Teraz preveďme decimálnu hodnotu na jej binárnu reprezentáciu a overme, že binárny ekvivalent $(65)_{10}$ je naozaj $(1000001)_2$. 

 
$$\text{Zdroj: NCERT (2019).}$$

V predošlej schéme sú vľavo samé dvojky, čo znamená, že vždy delíme kvocient číslom $2$ (chceme získať reprezentáciu daného čísla v dvojkovej sústave). Prostedný stĺpec obsahuje kvocienty po delení dvojkou (ako pri Euklidovom algoritme) a stĺpec napravo obsahuje zvyšky po delení číslom $2$.

## Prevod z dvojkovej sústavy do osmičkovej i šestnástkovej sústavy (a vice versa)

Číslo z dvojkovej sústavy (t.j. postupnosť bitov - núl a jednotiek) sa prevedie do osmičkovej alebo šestnástkovej sústavy nasledovne: Z daného binárneho čísla t.j. z postupnosti núl a jednotiek sa vytvoria trojice alebo štvorice bitov (t.j. núl a jednotiek) a každá skupinka bitov sa nahradí odpovedajúcou číslicou z osmičkovej sústavy (tzn. $0$-$7$) alebo odpovedajúcou číslicou zo šestnástkovej sústavy (tzn. $0$-$9$ a $A$-$F$). 

--- 

#### Poznámka 7 

* **Prečo sú trojice bitov v binárnom čísle zoskupované za účelom vytvorenia osmičkového čísla?** Základom osmičkovej sústavy je číslo $8$ a v osmičkovej sústave sa používa pri vytváraní čísel osem číslic, konkrétne $0$-$7$. Už vieme (pozri tabuľku vyššie - v časti osmičková sústava), že čím je číslo väčšie, tým viac bitov (núl a jednotiek) na jeho reprezentáciu potrebujeme. Najvyššia číslica v osmičkovej sústave je $7$ a zrejme $(7)_2=111$. Pre binárnu reprezentáciu  ostatných číslic osmičkovej sústavy teda určite nebude potrebných viac bitov než 3. Spočítajme všetky možné kombinácie troch binárnych číslic, tých je zrejme $2\cdot2\cdot2=2^3=8$, pretože na každej z troch pozícii máme dve možnosti, buď tam bude $1$ alebo $0$. Na reprezentáciu akejkoľvek číslice osmičkovej sústavy teda stačia 3 bity (číslic je osem, rôznych trojíc z núl a jedničiek je tiež osem). Preto sa vytvoria 3-bitové skupiny v binárnom čísle, pričom každá trojica bitov vyjadruje určitú číslicu v osmičkovej sústave a takto vznikne reprezentácia zadaného (binárneho) čísla v osmičkovej sústave. Na niektoré číslice osmičkovej sústavy ako napr. $3$, sú nutné len dva bity na jej binárnu reprezentáciu, lebo $(3)_2=11$ tzn. $1\times2^1+1\times2^0=3$, avšak nič nepokazíme, keď pred jednotky vsunieme nulu t.j. $(3)_2=011$, lebo platí $0\times2^2+1\times2^1+1\times2^0=3$. Rovanký princíp je možné uplatniť aj pri ďalších podobných prípadoch binárnej reprezentácie číslic osmičkovej sústavy. 


* Analogická úvaha platí aj pre reprezentáciu číslic šestnástkovej sústavy (t.j. $0$-$9$ a $A$-$F$) pomocou štvoríc bitov.

## Prevod čísla so zlomkovou časťou 

Zjednodušene povedané, **zlomková časť** reálneho čísla je všetko za desatinnou čiarkou vpravo. Tak napr. zlomková časť z čísla $123.45$ je $0.45$. Formálne by sme to zapísali takto: $\{123.45\}=123.45-\lfloor123.45\rfloor=123.45-123=0.45$. Zlomková časť reálneho čísla sa značí v množinových zátvorkach, tzn. zlomková časť z čísla $123.45$ je označená ako $\{123.45\}$, ďalej $\lfloor123.45\rfloor$ vyjadruje tzv. dolnú celú časť z reálneho čísla, v tomto prípade z čísla $123.45$, inými slovami je to najväčšie celé číslo, nie väčšie ako dané číslo, v tomto prípade je to $123$. Pri záporných číslach by mal byť človek opatrný lebo napr. $\lfloor-1.5\rfloor=-2$. 

Vo všeobecnosti by sme teda zlomkovú časť reálneho čísla $x$ mohli definovať nasledovne: 

$$\{x\}=x-\lfloor x \rfloor,$$ 

pričom číslo $x$ uvažujeme vo všeobecnosti v tvare $x=d_nd_{n-1}\ldots d_2d_1d_0.d_{-1}d_{-2}\ldots$, kde $d_n,d_{n-1},\ldots,d_1,d_0,d_{-1},d_{-2}\ldots$ sú číslice, ďalej $d_nd_{n-1}\ldots d_2d_1d_0$ zodpovedá celočíselnej časti reálneho čísla $x$ a $0.d_{-1}d_{-2}$ je zlomková časť reálneho čísla $x$. 

Doteraz sme sa do značnej miery zaoberali rôznymi prevodmi pre celé čísla. V tejto časti sa dozvieme o prevode čísel so zlomkovou časťou.  


--- 

### <font color=red> Definícia 2 </font>

Nech $b$ je prirodzené číslo väčšie ako $1$ a nech $n_0,n_1,\ldots,n_{b-1}$ je $b$ rôznych prirodzených čísel (nazývaných tiež číslice) takých, že $n_i$ označuje číslo $n_i=i$, pričom $i$ je z množiny $\{0,1,2,\ldots, b-1\}$. Reprezentácia zlomkovej časti čísla v číselnej sústave so základom $b$ je konečná alebo nekonečná (usporiadaná) postupnosť číslic $(0.d_{-1}d_{-2}d_{-3}\ldots)_d$, ktorá je interpretovaná ako reálne číslo nasledovným spôsobom 

$$d_{-1}b^{-1}+d_{-2}b^{-2}+d_{-3}b^{-3}+\ldots,$$

kde každá cifra $d_i$ vo vyjadrení čísla je rovná jednej z $b$ číslic $\{n_0,n_1,\ldots,n_{b-1}\}$, z ktorých číselná sústava so základom $b$ pozostáva.

 # <font color=brown> Aritmetické operácie v číselnej sústave so základom $b$ </font>

Metódy, pomocou ktorých sa v škole učíme vykonávať aritemetické operácie (napr. sčítanie, odčítane, násobenie, ...), sú úzko späté s vlastnosťami desiatkovej číselnej sústavy, avšak tieto metódy je možné ľahko zovšeobecniť resp. rozšíriť na akúkoľvek číselnú sústavu. Nebudeme to robiť podrobne, ale niekoľko príkladov by malo ilustrovať všeobecné (základné) myšlienky. Všetky metódy by mali byť známe zo školy, ale ak ste aritmetickým metódam nikdy celkom nerozumeli, možno budete musieť dvakrát premýšľať, aby ste pochopili, prečo to všetko funguje :-) Napriek tomu, že samotné metódy sú na celom svete rovnaké, treba spomenúť, že existuje veľa variácií v tom, ako sú metódy vyjadrené (zapísané) na papieri. Tu uvedený opis preto môže byť spočiatku neznámy a bude potrebné si na neho zvyknúť. **Myšlienky** a **postupy**, ktoré budeme ďalej prezentovať sú **aplikovateľné v ľubovoľnej číselnej sústave so základom $b$**. Ilustrovať budeme na príkladoch z osmičkovej alebo štvorkovej sústavy t.j. pre $b=8$ alebo $b=4$. 

## Sčítanie

Sčítanie dvoch jednociferných čísel je rovnaké ako v desiatkovej sústave, pokiaľ má výsledok iba jednu číslicu. Máme napríklad $4_8 + 3_8 = 4 + 3 = 7 = 7_8$. Ak výsledok sčítania v sústave so základom $b$ vyžaduje dve číslice, musíme si uvedomiť, že prechod je nie cez desiatku ale vo všeobecnosti cez základ $b$. Ak teda výjde výsledok rovný hodnote $b$ alebo väčší, tak výsledok bude mať dve číslice, kde číslica úplne vľavo je $1$ a tá druhá bude mať hodnotu súčtu, zníženú o základ $b$. To znamená, že 

$$5_8+6_8=5+6=11=8+11-8=8+3=1\times8^1+3\times8^0=13_8.$$ 

To sa dá napísať úplne rovnako, ako by sme zapísali súčet v desiatkovej sústave, musíme mať však na pamäti, že hodnota prenosu je $b$ (tzn. robíme sčítanie s prechodom nie cez desiatku ale vo všeobecnosti cez hodnotu $b$). Podobný príklad v desiatkovej sústave je možné zapísať nasledovne: 

$$8_{10}+9_{10}=8+9=17=10+17-10=10+7=1\times10^1+7\times10^0=17_{10}.$$ 

Skúsme teraz väčší súčet ako napr. $457_8 + 325_8$. To si vyžaduje postupné jednociferné sčítanie, rovnako ako v desiatkovej  sústave. Jeden zo spôsobov, ako to zapísať, je 

 

Predchádzajúci výpočet si rozpíšeme trochu podrobnejšie, aby sme lepšie videli, čo sa deje. V podstate ide o klasické sčítanie dvoch čísel pod sebou ale s prechodom cez osmičku (miesto desiatky), keďže teraz $b=8$. Pri sčítaní postupujeme štandardne sprava doľava (a zdola nahor) a využívame pritom sčítanie dvoch jednociferných čísel v osmičkovej sústave vysvetlené v príklade vyššie. Tzn. sčítavame postupne $5_8+7_8=5+7=12=8+12-8=8+4=1\times8^1+4\times8^0=14_8$, odkiaľ číslicu najviac vpravo t.j. číslicu $4$ napíšeme pod čiaru (do prvého stĺpca) a jednotku si poznačíme do druhého stĺpca hore (nad päťku). Ďalej sa posunieme o jednu pozíciu vľavo a sčítavame v druhom stĺpci t.j. $2_8+5_8+1_8$ avšak pripísali sme si tam i jendotku, ktorá sa "zvýšila" z predošlého kroku, máme teda $2_8+5_8+1_8=2+5+1=8=8+8-8=8+0=1\times8^1+0\times8^0=10_8$ a postup opakujeme, tzn. z čísla $10_8$ si číslicu $0$ napíšeme pod čiaru (dodruhého stĺpca, vedľa štvorky) a jednotku si poznačíme do tretieho stĺpca hore (nad štvorku). Opäť sa posunieme o jednu pozíciu vľavo a sčítavame v treťom stĺpci $3_8+4_8+1_8=3+4+1=8=8+8-8=8+0=1\times8^1+0\times8^0=10_8$ a toto číslo už celé napíšene pod čiaru (keďže sme už v poslednom stĺpci pre sčítavanie), takto dostaneme výsledok sčítania dvoch čísel v osmičkovej sústave.

Tomuto zodpovedá sčítanie v desiatkovej sústave $303 + 213 = 516$. 

```python
# kontrola v Sage 
x = Integer('457', base=8) # zadame cislo x v osmickovej sustave 
show(x.str(base=8)) # vypiseme cislo x v osmickovej sustave 
show(x) # vypiseme cislo x v desiatkovej sustave 
y = Integer('325', base=8) # zadame cislo y v osmickovej sustave 
show(y.str(base=8)) # vypiseme cislo y v osmickovej sustave 
show(y) # vypiseme cislo y v desiatkovej sustave 
z = x + y # sucet x+y
show(z.str(base=8)) # vypiseme sucet x+y v osmickovej sustave 
show(z) # vypiseme sucet x+y v desiatkovej sustave 
```

## Odčítanie 

Jednociferné odčítania sú jednoduché, napríklad $7_8 −3_8 = 4_8$. Odčítanie ako $14_8 −7_8$ (s "prechodom" cez osmičku) je o niečo ťažšie, ale môžeme si "požičať" z "$1$" v $14$ rovnako ako v desiatkovej sústave. Jediným rozdielom je, že pri základe $8$, jednotka "$1$"  predstavuje hodnotu $8$ a nie $10$, takže si "požičiavame" v skutočnosti $8$. Potom vidíme, že musíme vykonať odčítanie $12–7$, takže výsledok je $5$ (v desiatkovej sústave i v osmičkovej sústave). Odčítanie väčších čísel je možné vykonať opakovaním tohto postupu. Uvažujme napríklad $321_8 - 177_8$. To sa dá napísať v tvare 

 

Opäť je postup analogický ako pri odčítavaní dvoch čísel (s prechodom cez desiatku) pod sebou v desiatkovej sústave ale treba dať pozor na to, že teraz "neprechádzame" cez desiatku ale cez osmičku! Tzn. výpočet na obrázku si môžeme rozpísať takto: Začíname v stĺpci úplne vpravo a keďže $1<7$ pýtame sa, "$7$ a koľko je $9$ ?"; odpoveď je $2$, čo zapíšeme pod čiaru do prvého stĺpca. Všimnme si, keďže $1<7$, tak sme si "požicali" osmičku a pripočítali ju k $1$, preto $1+8=9$. Ak by sme boli v desiatkovej sústave, pýtali by sme sa "$7$ a koľko je $11$ ?" (lebo $1+10=11$, t.j. "požičali" by sme si desiatku). "Požicanú" osmičku si zapíšeme nad jednotku do pravého krajného stĺpca a posunieme sa o jednu pozíciu vľavo k prostrednému stĺpcu. K sedmičke pripočítame jednotku (zodpovedá "požicanej" osmičke z predošlého kroku) t.j. $7+1=8$ a pýtame sa, "$8$ a koľko je $10$ ?" (opať sme si "požicali" osmičku lebo $2<8$). Odpoveď je $2$, čo zapíšeme pod čiaru. "Požičanú" osmičku poznačíme nad dvojku v prostrednom stĺpci. Následne sa presunieme do posledného (ľavého krajného) stĺpca, kde k jednotke pripočítame jednotku (zodpovedá "požičanej" osmičke z predošlého kroku) a pýtame sa, "$2$ a koľko je $3$ ?" Odpoveď je jedna, čo zapíšeme pod čiaru. Takto dostaneme výsledok odčítania dvoch čísel v osmičkovej sústave. 

```python
# kontrola v Sage 
x = Integer('321', base=8) # zadame cislo x v osmickovej sustave 
show(x.str(base=8)) # vypiseme cislo x v osmickovej sustave 
# show(x) # vypiseme cislo x v desiatkovej sustave 
y = Integer('177', base=8) # zadame cislo y v osmickovej sustave 
show(y.str(base=8)) # vypiseme cislo y v osmickovej sustave 
# show(y) # vypiseme cislo y v desiatkovej sustave 
z = x - y # rozdiel x-y
show(z.str(base=8)) # vypiseme rozdiel x-y v osmickovej sustave 
# show(z) # vypiseme rozdiel x-y v desiatkovej sustave 
```

<font color=orange>  Dôkladným rozpísaním  preveďte príklad (všetky čísla v ňom) $321_8 - 177_8 = 122_8$  do desiatkovej sústavy a tam ho vypočítajte. Svoje riešenie (prevody aj sčítanie) skontrolujte v Sage.  </font>

## Násobenie 

Vezmime si jeden príklad násobenia, konkrétne $312_4 \times 12_4$. Podobne ako v desiatkovej sústave je základom pre násobenie viacciferných čísel násobiaca tabuľka pre jednociferné čísla. Pre základ $4$ je multiplikačná tabuľka nasledovná (čísla v tabuľke sú uvedené pri základe $4$)

 

Násobenie potom môžeme vykonať tak, ako sme na to zvyknutí v desiatkovej sústave 

 

Číslo $1230_4$ v druhom riadku je výsledkom násobenia $312_4 × 2_4$, t.j. prvý činiteľ $312_4$ vynásobený druhou číslicou (najviac vpravo) druhého činiteľa $12_4$. Číslicou $2$ sa násobí číslo $1230_4$ postupne sprava doľava a využíva sa pri tom tabuľka uvedená vyššie pre násobenie v štvorkovej sústave, tzn. $2_4\times2_4=10_2$, odkiaľ $0$ zapíšeme a $1$ si zapamätáme, ďalej násobime $2_4\times1_4=2_4$ ale ešte pripočítame $1$ z predošlého kroku t.j. $2_4+1_4=3_4$ a teda trojku zapíšeme vedľa nuly vľavo, ďalej násobime $2_4\times3_4=12_4$ a keďže je to posledné násobenie v rade, tak obe číslice $12$ zapíšeme do výsledku, čím dostaname $312_4 × 2_4=1230_4$. 
Číslo $3124$ v riadku nižšie, je prvý činiteľ $312_4$ vynásobený prvou číslicou $1_4$ (najviac vľavo) druhého činiteľa $12_4$. Tento súčin sa určí analogicky ako ten predchádzajúci. Dostaneme tak $312_4\times1_4=312_4$ a tento výsledk súčinu je posunutý o jedno miesto doľava, pretože vynásobenie prvou číslicou v $12_4$ zodpovedá násobeniu $1 \times 4$. Číslo $11010_4$ v poslednom riadku je súčtom dvoch vyššie uvedených čísel ($1230_4$ a $312_4$), pričom na pravom konci čísla $312_4$ je pridaná  nula, t.j. súčet je $1230_4 + 3120_4$. Tento súčet sa vypočíta podľa pravidiel uvedených v odseku o sčítaní.
