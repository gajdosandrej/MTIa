# <font color=brown> Motivácia </font>

V rámci motivácie k téme 9 sme stručne poukázali na dôležitosť **matematických funkcií** a ich uplatnenie v niektoých oblastiach informačných technológii (je to len zlomok toho, kde všade sú funkcie potrebné). Keď už človek pracuje v konkrétnom odvetví, rieši praktické problémy/úlohy a narazí v aplikáciách na určité funkcie, potrebuje poznať ich vlastnosti, aby s nimi mohol správnym spôsobom zaobchádzať. Napr. taký **vývojár videohier**, ktorý má na starosti **fyzikálny "engine" hry** (funkcie a rovnice), tzn. je zodpovedný za **realistickú simuláciu fyzikálnych zákonov** v rámci hry. Fyzikálny "engine" hry poskytuje sadu funkcií na simuláciu fyzikálnych síl a zrážok, pôsobiacich na rôzne objekty v hre v reálnom čase. Cieľom je, aby hra a všetko čo sa v nej pohybuje alebo nejakým spôsobom mení, sa čo najviac podobalo reálnemu svetu. 



V téme 10 si teda predstavíme (pripomenieme) **základe typy (triedy) funkcií** tzv. **elementárne funkcie** a ich vlastnosti. Konkrétne sa oboznámime s **polynomickými** a **mocninovými funkciami**, s **exponenciálnymi** a **logaritmickými funkciami**, ale aj s **goniometrickými funkciami**. Dôležité bude uvedomiť si, **kde všade sú a kde nie sú konkrétne funkcie definované**, kde a za akých podmienok sú **rastúce či klesajúce**, ak majú **periódu** tak akú a pod. Dôležitým aspektom bude tiež uvedomiť si a porozumieť **ako sa menia grafy** daných **funkcií** po vynásobení konštantou, po pripočítaní konštanty atď., resp. aj opačný proces, ak máme zadanú funkciu, **ako** z nej **vytvoriť funkciu** rovnakého typu ale s niektorými odlišnými vlastnosťami?
