#### <font color=blue> Príklad 1 </font> 

Predpokladajme, že spoločnosť zaoberajúca sa počítačovým softvérom vyrába a predáva nový operačný systém za cenu 55 dolárov za kópiu a spoločnosť má fixné náklady 15 000 dolárov mesačne. Celkové mesačné náklady vyjadrite v závislosti od počtu predaných kópií $x$ a vypočítajte náklady, keď x = 1000. 

<ins> Riešenie: <ins/> 
    
Premenlivé (pohyblivé) mesačné výdavky, závisiace od počtu predaných kópii softvéru sú $55x$, kde $x$ je počet predaných kópií softvéru. Teda celkové mesačné výdavky spoločnosti sú fixné mesačné výdavky plus mesačné výdavky spojené s predaným softvérom, t.j. 

$$C(x)=15000+55x.$$ 

Ak sa predá $1000$ kópií operačného systému zamesiac, tak celkové mesačné náklady firmy sú 

$$C(1000)=15000+55\cdot1000=70000\,\text{( dolárov)}.$$

```python
x =  var('x')
C(x) = 15000 + 55*x 
kwds = {'fontsize':12} 
plot(C(x), (x, 0, 500), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2) + text(r"$y=C(x)$", (250, 35000), color='blue',**kwds) + text("Predané kópie (mesačne)", (250, -6000), color='black') + text("Náklady na predaný softvér", (-100, 25000), color='black', rotation=90.0)
```

### <font color=red> Definícia 1 </font> 

Funkcia $f:\ \mathbb{R} \rightarrow \mathbb{R}$ daná predpisom $f(x) = a_0 + a_1x + a_2x^2 + \dots + a_n x^n$, kde $n \in \mathbb{N}_0,\,a_0, a_1, \dots , a_n \in \mathbb{R}$ sa nazýva **_polynomická_** (výraz $a_0 + a_1x + a_2x^2 + \dots + a_n x^n$ je **_polynóm_**). Ak $f$ je nenulová a $a_n \neq 0$, tak $n$ sa volá **_stupeň_** polynomickej funkcie.

Ak $n=0$, tak $f$ sa nazýva **_konštantná_** funkcia. Ak stupeň funkcie $f$ je $n=1$ (2, 3), tak $f$ sa nazýva **_lineárna_** (**_kvadratická_**, **_kubická_**) funkcia.

#### <font color=blue> Príklad 3 </font> 

Racionálne funkcie sa používajú v environmentálnych štúdiách ako modely nákladov a úžitkov. Náklady na odstránenie znečisťujúcej látky z atmosféry sa odhadujú ako funkcia percentuálneho podielu odstránenej znečisťujúcej látky. Čím vyššie je odstránené percento, tým väčší je „úžitok“ pre ľudí, ktorí dýchajú tento vzduch. Tieto problémy sú, samozrejme, komplexné a definícia „nákladov“ je diskutabilná. Náklady na odstránenie malého percenta znečisťujúcich látok môžu byť dosť nízke. Avšak na druhej strane, náklady na odstránenie konečných 5% znečisťujúcej látky môžu byť veľmi vysoké. 

Nasledujúca funkcia je príkladom modelu popísaného vyššie: 

$$f(x)=\dfrac{50x}{105-x},\qquad 0\le x \le 100,$$ 

kde $x$ je percento znečisťujúcej látky, ktoré sa má odstrániť a $f(x)$ sú tomu odpovedajúce náklady (na odstránenie, v miliónoch dolárov). 

Nájdite náklady na odstránenie $70%$, $95%$ a $100%$ znečisťújúcej látky. 

<ins> Riešenie: <ins/> 

Náklady na odstránenie $70%$ znečisťujúcej látky sú 

$$f(70)=\dfrac{50\cdot70}{105-70}=100\,\text{miliónov dolárov}.$$ 

Podobne určíme aj ďalšie náklady 

$$f(95)=475\qquad\qquad \text{ a } f(100)=1000.$$ 

Všimnime si, že náklady na odstránenie posledných $5%$ znečisťujúcej látky sú $f (100) - f (95) = 1 000 - 475 = 525$ miliónov dolárov. To je viac ako päťnásobok nákladov na odstránenie prvých $70%$ znečisťujúcej látky!


```python
x =  var('x')
f(x) = 50 * x / (105 - x) 
kwds = {'fontsize':12} 
plot(f(x), (x, 0, 104), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2) + text(r"$y=f(x)$", (90, 5000), color='blue',**kwds) + text(r"Percento odstránenej znečisťujúcej látky", (50, -700), color='black') + text("Náklady na odstránenie (v miliónoch dolárov)", (-15, 3000), color='black', rotation=90.0) + line([(105,0),(105,6000)], color = 'red') + text("$x=105$", (115, 3000), color='red',**kwds) 
```

#### Poznámka 2 

Ak by sme v Príklade 3 uvažovali funkciu $f(x)=\dfrac{50x}{105-x}$ aj mimo interval $\langle 0,100\rangle$ a dosadzovali by sme postupne do funkcie $f(x)$ také $x$, ktoré sú stále bližšie a bližšie k hodnote $105$, tak funkcia $f(x)$ by nadobúdala stále väčšie a väčšie hodnoty. Inými slovami, ak sa $x$ približuje hodnote $105$, tak graf funkcie $f(x)$ sa v nekonečne približuje priamke $x=105$ (červená čiara v grafe), no nikdy sa jej nedotkne (viď graf v Príklade 3). Pre $x=105$ funkcia $f(x)$ nie je definovaná (nula v menovateli zlomku). Priamku $x=105$ voláme **_vertikálna asymptota_** funkcie $f(x)$, je to vlastne priamka rovnobežná s osou $y$. 

Podobne sa môže graf funkcie blížiť k priamke rovnobežnej s osou $x$, keď x postupne nadobúda veľmi veľké hodnoty (x ide do nekonečna) alebo veľmi malé hodnoty (x ide do mínus nekonečna). Takú priamku voláme **_horizontálna asymptota_**.

### <font color=red> Definícia 2 </font> 

Ak $P,Q$ sú polynomické funkcie a $Q$ je nenulová, tak funkcia $f = \dfrac{P}{Q}$ sa nazýva **_racionálna_** funkcia. 
		
Nenulová racionálna funkcia $\dfrac{P}{Q}$ sa volá **_rýdzoracionálna_**, ak stupeň $P$ je menší ako stupeň $Q$.

--- 

Konkrétne v Príklade 3 sme mali $P=50x$ a $Q=105-x$. 

**Podobne ako v prípade celých čísel existuje Veta o delení so zvyškom, aj pre polynomické funkcie platí analogické tvrdenie.** 

---

#### <font color=green> Tvrdenie 1 </font>

Nech $P,Q$ sú polynomické funkcie, $Q$ je nenulová. Potom existuje jediná dvojica polynomických funkcií  $U,V$ taká, že $P = U\cdot Q + V$, pričom buď $V$ je nulová funkcia, alebo stupeň $V$ je menší  ako stupeň $Q$.

---

Pomocou Tvrdenia 1 je možné dokázať nasledujúcu Vetu o reprezentácii racionálnych funkcií. 

---

### <font color=green> Veta 1 </font>

Každá racionálna funkcia sa dá vyjadiť jedným z nasledujúcich dvoch spôsobov:

1. je rovná polynomickej funkcii s definičným oborom rovným definičnému oboru danej funkcie;


2. je rovná súčtu polynomickej a rýdzoracionálnej funkcie.

#### <font color=blue> Príklad 4 </font>

Nech $P(x) = x^5 - 5x^4 - 2x^3 + 12x^2 - 2x+12,\, Q(x) = x^3 - 5x^2-3x+17$. Nájdite funkcie $U,V$ tak, aby $P(x) = U(x)\cdot Q(x) + V(x)$ pre každé $x \in \mathbb{R}$.

<ins> Riešenie: <ins/> 

Vypočítame podiel polynómov $P$ a $Q$: 



Potom $U$ je čiastočný podiel, $U(x) = x^2+1$ a $V$ je zvyšok po delení, $V(x) = x-5$. 

Platí teda $P(x) = U(x)\cdot Q(x) + V(x)$ t.j. $x^5 - 5x^4 - 2x^3 + 12x^2 - 2x+12 = (x^2+1)\cdot (x^3 - 5x^2-3x+17) + (x-5)$. 

```python
# kontrolny vypocet v Sage
x = var('x')
P(x) = x^5-5*x^4-2*x^3+12*x^2-2*x+12
Q(x) = x^3-5*x^2-3*x+17
U,V = P.maxima_methods().divide(Q) # delenie polynomov v Sage 
show(LatexExpr('P(x) = ') + latex(P(x)))
show(LatexExpr('Q(x) = ') + latex(Q(x)))
show(LatexExpr('U(x) = ') + latex(U(x)))
show(LatexExpr('V(x) = ') + latex(V(x)))
# kontrola 
show(LatexExpr('U(x)\\cdot Q(x) + V(x) = ') + latex(expand(U*Q+V)(x)) + LatexExpr(' = P(x)'))
```

Vydeľme populáciu v nejakom roku populáciou v predošlom roku, napr.: 

$$\dfrac{\text{Populácia v roku 2008}}{\text{Populácia v roku 2007}}=\dfrac{14.660 \text{ miliónov}}{14.235 \text{ miliónov}}=1.029, $$

$$\dfrac{\text{Populácia v roku 2009}}{\text{Populácia v roku 2008}}=\dfrac{15.095 \text{ miliónov}}{14.660 \text{ miliónov}}=1.030.$$ 

Skutočnosť, že obidva výpočty dávajú podiel približne $1.03$, ukazuje, že populácia vzrástla o približne $3\%$ v rokoch 2008 až 2009 a medzi rokmi 2007 a 2008. Analogické výpočty za ostatné roky ukazujú, že populácia rástla faktorom (koeficientom) približne $1.029$ resp. $2,9%$ každý rok. Kedykoľvek máme **konštantný rastový faktor** (tu $1.029$), máme exponenciálny rast. Populácia $t$ rokov po roku 2007 je daná exponenciálnou funkciou 

$$P=14.235\cdot(1.029)^t.$$ 

Ak predpokladáme, že vzorec platí 50 rokov, graf populácie má tvar znázornený na obrázku pod textom. Keďže počet obyvateľov rastie rýchlejšie a rýchlejšie, graf sa ohýba nahor; hovoríme, že je _konkávny_. Dokonca aj exponenciálne funkcie, ktoré spočiatku stúpajú pomaly, ako je táto, nakoniec stúpajú veľmi rýchlo. 

Pozn.: Na to, aby sme z tabuľky hodnôt pre $t$ a $P$ zistili, že pochádzajú z exponenciánej funkcie, stačí sa pozrieť na podiely hodnôt $P$ (odpovedajúce príslušným $t$); tieto podiely musia byť konštantné pre rovnako vzdialené hodnoty $t$ (viď. príklad s tabuľkou vyšsie).  

```python
t =  var('t')
P(t) = 14.235 * (1.029 ^ t)
kwds = {'fontsize':12} 
plot(P(t), (x, -10, 50), gridlines = 'minor', axes_labels=[r'$t$ (roky od 2007)', r'$P$ (populácia v miliónoch obyvateľov)'], axes_labels_size = 1) + text(r"$P=14.235\cdot(1.029)^t$", (50, 65), color='blue',**kwds, figsize=5) 
```

Zlogaritmujeme obe strany vyššie uvedenej rovnice (môžeme to urobiť za predpokladu, že obe strany rovnice sú kladné, v opačnom prípade logaritmická funkcia nie je definovaná), čím dostávame  

$$\log P = \log \left(14.235\cdot(1.029)^t\right).$$

Daľej, keďže podľa vlastností logaritmickej funkcie platí $\log\left(AB\right)=\log A + \log B$, máme 

$$\log P=\log 14.235 + \log(1.029)^t.$$

Teraz využijeme vlastnosť $\log\left(A^t\right)=t\log A$, čo nám dá 

$$\log P = \log 14.235 + t\log1.029.$$

Teraz je už rovnica lineárna vzhladom na premennú $t$, čize vyjadríme $t$ nasledovne: 

$$t\log 1.029=\log P - \log 14.235,$$

$$t=\dfrac{\log P}{\log 1.029}-\dfrac{\log 14.235}{\log 1.029}=80.545\log P-92.898.$$

Preto 

$$t=f^{-1}(P)=80.545\log P-92.898.$$

$$t=f^{-1}(20)=80.545(\log 20)-92.898=11.89;$$

 # <font color=brown> Goniometrické funkcie</font>

Trigonometria vznikla ako súčasť štúdia trojuholníkov a prvé definície goniometrických (trigonometrických) funkcií boli v trojuholníkoch. Goniometrické funkcie sa však dajú definovať aj pomocou jednotkovej kružnice, čo je definícia, ktorá z nich robí periodické alebo opakujúce sa funkcie. Mnoho prirodzene sa vyskytujúcich procesov je periodických. Hladina vody v prílivovej nádrži, krvný tlak v srdci, striedavý prúd či poloha molekúl vzduchu prenášajúcich hudobný tón pravidelne kolíšu. Takéto javy môžu byť reprezentované goniometrickými funkciami. 

Existujú dva bežne používané spôsoby reprezentácie vstupu trigonometrických funkcií: **_radiány_** a **_stupne_**. Ako uvidíme, vzorce kalkulu sú v radiánoch elegantnejšie ako v stupňoch. 

Uhol $1$ **_radián_** je definovaný ako uhol v strede jednotkovej kružnice, ktorý vymedzuje oblúk s dĺžkou $1$, meraný proti smeru hodinových ručičiek (pozri obrázok (a) nižšie). **_Jednotková kružnica_** má polomer $1$ (a stred v začiatku súradnicovej sústavy, t.j. v bode $[0,0]$). 

Uhol $2$ radiány vymedzí oblúk s dĺžkou $2$ na jednotkovej kružnici. Záporný uhol, ako napríklad $-1/2$ radiánu, vymedzí oblúk s dĺžkou $1∕2$, ale meraný v smere hodinových ručičiek. (Pozri obrázok (b) nižšie). 

 
$$\text{Zdroj: Hallet (2017).}$$ 

Je užitočné vnímať uhly ako rotácie (otočenia), potom môžeme dať zmysel uhlom väčším ako $360^\circ$; napríklad uhol $720^\circ$ predstavuje dve úplné (celé) otáčky (otočenia) proti smeru hodinových ručičiek. Pretože jedna úplná rotácia o $360^\circ$ vymedzí oblúk s dĺžkou $2\pi$, čo je vlastne obvod jednotkovej kružnice ($2\pi\cdot1$), z toho vyplýva, že 

$$360^\circ=2\pi \text{ radiánov}\qquad\qquad \text{ a z toho }\qquad\qquad 180^\circ=\pi\text{ radiánov}.$$ 

Inými slovami, $1$ radián $= 180^\circ/\pi$, takže jeden radián je približne $60^\circ$. Slovo radián(y) sa často vynecháva, takže ak sa na uhol alebo rotáciu (otočenie) odkazuje bez jednotiek, chápe sa to v radiánoch. 

Radiány sú užitočné na výpočet dĺžky oblúka v ľubovoľnom kruhu. Ak má kruh polomer $r$ a oblúk vymedzí uhol $\theta$ (ako na ďalšom obrázku), potom platí nasledujúci vzťah: 

$$\text{dĺžka oblúka } = s = r\cdot\theta.$$ 

 
$$\text{Zdroj: Hallet (2017).}$$ 

Všimnime si, že špeciálne pre jednotkovú kružnicu (kde polomer $r=1$) platí $s=\theta$ t.j. uhol o velkosti $\theta$ radiánov vymedzí na kružnici oblúk dĺžky $\theta$ (jednotka dĺžky teraz nie je podstatná, môžu to byť napr. cm alebo m). 

Pojem dĺźka oblúka sa zvykne nazývať aj **_oblúková miera_** prislúchajúca určitému uhlu $\theta$. Formálne to zhŕňa nasledujúca definícia. 

--- 

### <font color=red> Definícia 6 </font> 

**_Oblúkovou mierou_** uhla $\theta$ rozumieme dĺžku oblúka, ktorý vytínajú ramená stredového uhla $\theta$ na jednotkovej kružnici; vrchol uhla $\theta$ je totožný so stredom jednotkovej kružnice, prvé rameno uhla $\theta$ je totožné s polpriamkou $x \geq 0,y=0$ a orientácia $\theta$ (v smere od prvého ramena k druhému) je v kladnom zmysle, t.j. proti smeru hodinových ručičiek. 

--- 

Predpokladajme ďalej, že na jednotkovej kružnici sa nachádza bod $A$ so súradnicami $x_A,y_A$, t.j. $A=[x_A,y_A]$ taký, že $A$ je priesečník jednotkovej kružnice a druhého ramena uhla s oblúkovou mierou $x$ (pred tým označená na obrázku ako $s$), tzn. uhol veľkosti $x$ radiánov (uhol odpovedajúci dĺžke oblúka $x$ dĺžkových jednotiek). Potom jednoducho definujeme funkcie $\sin x$ a $\cos x$ ako súradnice bodu $A$ pohybujúceho sa po jednotkovej kružnici a vymedzujúceho na tejto kružnici oblúk dlžky $x$  takto: $\cos x=x_A$ a $\cos x = y_A$, ako to znázorňujú i nasledujúce obrázky. 

 
$$\text{Zdroj: Madaras (2019).}$$

--- 

### <font color=red> Definícia 7 </font> 

Nech $x \in \langle 0 , 2\pi)$ a nech $A=[x_A,y_A]$ je priesečník jednotkovej kružnice a druhého ramena uhla s oblúkovou mierou $x$. Potom hodnota $x_A$ sa označuje $\cos x$ a $y_A$ sa označuje $\sin x$.

--- 

--- 

### <font color=red> Definícia 8 </font> 

Periodickú funkciu $f:\ \mathbb{R} \to \mathbb{R}$ s periódou $2\pi$ s vlastnosťou, že pre každé $x_0 \in \langle 0 ,2\pi)$ platí $f(x_0) = \sin x_0$, nazývame **_sínus_** $x$ a označujeme $f:\ y = \sin x$. Podobne, periodickú funkciu $g:\ \mathbb{R} \to \mathbb{R}$ s periódou $2\pi$ s vlastnosťou, že pre každé $x_0 \in \langle 0 ,2\pi)$ platí $g(x_0) = \cos x_0$, nazývame **_kosínus_** $x$ a označujeme $g:\ y = \cos x$.

--- 

Keďže pre ľubovoľný bod $A=[x_A,y_A]$ na jednotkovej kružnici platí Pytagorova veta tak dostávame rovnicu jednotkovej kružnice  $x_A^2+y_A^2=1$ a po nahradení máme známy vzťah (uvedený aj nižšie medzi vzorcami)

$$\cos^2x+\sin^2x=1.$$ 

Keď sa $x$ (dĺžka oblúka resp. prejdená vzdialenosť bodom $A$ po kružnci) zvyšuje a bod $A$ sa pohybuje po kružnici, hodnoty $\sin x$ a $\cos x$ sa hýbu medzi medzi $1$ a $-1$ a opakujú sa, keď sa bod $A$ pohybuje cez body, kde už bol predtým. Ak je $x$ záporné, uhol sa meria v smere hodinových ručičiek po kružnici. 

  
$$\text{Zdroj: Madaras (2019).}$$ 


$$\text{Zdroj: Hallet (2017).}$$

Grafy sínusu a kosínu sú znázornené na predchádzajúcom obrázku. Všimnime si, že sínus je nepárna funkcia a kosínus je párny. Maximálna a minimálna hodnota sínusu a kosínu sú $+1$ a $-1$, pretože to sú maximálne a minimálne hodnoty (súradníc) $y$ a $x$ na jednotkovej kružnici. Po tom, čo už bod $A$ raz prešiel celú kružnicu, začnú sa hodnoty $\cos x$ a $\sin x$ opakovať; hovoríme, že funkcie sú periodické. 

Amplitúda $\cos x$ a $\sin x$ je $1$ a perióda je $2$. Prečo $2$ ? Pretože toto je hodnota $x$ (dĺžka resp. obvod jednotkovej kružnice), keď bod $A$ prešiel presne jedenkrát celú kružnicu. (Nezabudnite, že $360^\circ = 2 $ radiány.) 

Na obrázku tiež vidíme, že sínusový a kosínusový graf majú presne rovnaký tvar, iba horizontálne posunutý. Pretože kosínusový graf je sínusový graf posunutý o $\pi/2$ doľava, tak platí 

$$\cos x = \sin(x+\pi/2).$$ 

Rovnako platí, že sínusový graf je vlastne kosínusový graf posunutý o $\pi/2$ doprava, t.j. 

$$\sin x = \cos (x-\pi/2).$$ 

Hovoríme, že **_fázový rozdiel_** alebo **_fázový posun_** medzi $\sin x$ a $\cos x$ je $\pi/2$. 

Funkcie, ktorých grafy majú tvar sínusovej alebo kosínusovej krivky, sa nazývajú **_sínusoidalné funkcie._** 

Na opis ľubovoľných amplitúd a periód sínusoidálnych funkcií používame funkcie v tvare 

$$f(x)=A\sin(Bx)\qquad\qquad\text{ a }\qquad\qquad g(x)=A\cos(Bx),$$

**kde $|A|$ je amplitúda a $2\pi/|B|$ je perióda**. <br/>
Ak argument $x$ v predpise sínusoidálnej funkcie nahradíme argumentom $x+h$ alebo $x-h$, tak potom graf takej sínusoidálnej funkcie je horizontálne posunutý o hodnotu $|h|$. <br/> 
Funkcie v tvare $f(x)=A\sin(Bx)+C$ a $g(x)=A\cos(Bx)+C$ majú grafy vertikálne posunuté o hodnotu $C$ a tieto funkcie nadobúdajú hodnoty medzi $-1+C$ a $1+C$.

### <font color=red> Definícia 9 </font> 

Funkciu $h:\ y = \dfrac{\sin x}{\cos x}$ nazývame **_tangens_** $x$ a označujeme $h: y = \mathrm{tg}\, x$ (niekedy tiež $\mathrm{tan}\, x$). 

Funkciu $q:\ y = \dfrac{\cos x}{\sin x}$ nazývame **_kotangens_** $x$ a označujeme $q: y = \mathrm{cotg}\, x$.

--- 

--- 

#### Poznámka 6 

* Geometricky význam hodnoty funkcie $\mathrm{tg}\,x$ je ten, že vyjadruje sklon priamky prechádzajúcej stredom jednotkovej kružnice $[0,0]$ a bodom $A=[x_A,y_A]=[\cos x, \sin x]$ na jej obvode. 


* Funkcia tangens nie je definovaná v bodoch, kde $\cos x=0$, konkrétne ide o body z množiny $\{\pm\pi/2,\pm3\pi/2,\ldots\}$. 


* Funkcia $\mathrm{tg}\, x$ nadobúda kladné hodnoty (je kladná) tam, kde $\sin x$ a $\cos x$ majú rovnaké znamienka.

Funkcia tangens má periódu $\pi$, pretože sa jej hodnoty opakujú každých $\pi$ jednotiek. Má zmysel hovoriť o amplitúde  funkcie tangens? Nie, ak uvažujeme o amplitúde ako o mierke veľkosti kmitania (výkyvov funkcie resp. rozpätí funkčných hodnôt), pretože funkcia tangens sa stáva nekonečne veľká (hodnota funkcie rastie do nekonečna) pri každej vertikálnej asymptote.  # <font color=brown> Cyklometrické funkcie </font>

Niekedy pri riešení praktických úloh sa môžu objaviť rovnice nasledujúceho typu: 

$$\sin x=0,\qquad\qquad\sin x = 0.3.$$

Prvá z rovníc ma riešenia z množiny $\{0,\pm\pi,\pm2\pi,\ldots\}$, druhá rovnica má (približné resp. numerické) riešenia obsiahnuté v množine $\{0.305, 2.84, 0.305\pm2\pi, 2.84\pm2\pi,\ldots\}$. Riešení je v oboch prípadoch viac (až nekonečne veľa) kvôli tomu, že funkcia sínus (ako aj ostatné goniometrické funkcie) nie je prostá na celom svojom definičnom obore $\mathbb{R}$ (pozri napr. graf) ale je prostá napr. na intervale $\langle0,2\pi)$ alebo na intervale $\langle-\pi/2,\pi/2\rangle$. 

Môžeme vybrať pre každú z uvedených rovníc ako preferované riešenie napr. riešenie patriace do intervalu $\langle-\pi/2,\pi/2\rangle$. Tzn. podľa tejto dohody by preferované riešenie prvej rovnice bolo $x=0$ a preferované riešenie druhej rovnice by bolo $x=0.305$. Preferované riešenie sa niekedy zvykne označovať ako _hlavný argument funkcie_ (v tomto prípade funkcie sínus). 

Keďže goniometrické funkcie nie sú prosté, je nutné sa obmedziť na nejaký rozumný interval, na ktorom sú uvažované goniometrické prosté a tak môžeme na zvolených intervaloch definovať k nim inverzné funkcie. 

Vzhľadom k vyššie uvedeným úvahám sa zvykne definovať inverzná funkcia k funkcii sínus ako funkcia, ktorá dáva tzv. "preferované riešenie" resp. hlavný argument funkcie a označuje sa $\mathrm{arcsin}$ alebo aj $\sin^{-1}$  a nazýva sa **_arkussínus_**. <br/>
Teda pre $-1\le y \le 1$

$$\mathrm{arcsin}\,y=x,$$

znamená 

$$\sin x=y,$$

kde $-\dfrac{\pi}{2}\le x \le \dfrac{\pi}{2}$. 

Čiže arkussínus je inverzná funkcia k časti funkcie sínus definovanej na intervale $\langle-\pi/2,\pi/2\rangle$. 

Vysvetlime si ešte jeden (geometrický) pohľad na funkciu arkussínus resp. na to, ako ju zaviesť. Pri zavádzaní funkcie $\sin x$ nás zaujímala druhá súradnica bodu $A=[x_A,y_A]$, ktorý sa pohybuje po jednotkovej kružnci proti smeru hodinových ručičiek od bodu $[1,0]$ a prejde po kružnici (po oblúku) vzdialenosť $x$ (bod $A$ vymedzí na jednotkovej kružnici oblúk dĺžky $x$)., t.j. $\sin x = y_A$.  Môžeme však položiť otázku aj opačne: Ak je daná druhá súradnica bodu $A=[x_A,y_A]$ na jednotkovej kružnici, aká je odpovedajúca dĺžka oblúka (jednotkovej kružnice), ktorý vymedzuje bod $A$ ? Vďaka úvahám uvedeným vyššie už vieme, že správnych odpovedí je nekonečne veľa, ale ako _hlavnú hodnotu_ funkcie arkussínus môžeme vziať tú dĺžku oblúka, ktorá má najbližšie k východiskovému bodu $[1,0]$ na jednotkovej kružnici. Takto dostaneme $\mathrm{arcsin}\,y_A=x$. 

--- 

### <font color=red> Definícia 10 </font> 

Nech $g:\ y=\sin x$, $\mathcal{D}(g)= \langle -\frac{\pi}{2} , \frac{\pi}{2} \rangle$. Inverzná funkcia $f$ k funkcii $g$ sa  nazýva **_arkussínus_** a označuje sa  $\,f:\ y=\mathrm{arcsin}\, x.$ 

Nech $g:\ y=\cos x$, $\mathcal{D}(g)= \langle 0 , \pi\rangle$. Inverzná funkcia $f$ k funkcii $g$ sa  nazýva **_arkuskosínus_**  a označuje sa $\,f:\ y=\mathrm{arccos}\, x.$

---

### <font color=red> Definícia 11 </font> 

Nech $g:\ y=\mathrm{tg}\, x$, $\mathcal{D}(g)= \langle -\frac{\pi}{2} , \frac{\pi}{2} \rangle$. Inverzná funkcia $f$ k funkcii $g$ sa  nazýva **_arkustangens_** a označuje sa $\,f:\ y=\mathrm{arctg}\, x.$ 

Nech $g:\ y=\mathrm{cotg}\, x$, $\mathrm{D}(g)= \langle 0 , \pi \rangle$. Inverzná funkcia $f$ k funkcii $g$ sa  nazýva **_arkuskotangens_** a označuje sa $\,f:\ y=\mathrm{arccotg}\, x.$

--- 


$$\text{Zdroj: Madaras (2019).}$$

**Vzťahy a vzorce pre cyklometrické funkcie:**


* $\mathrm{arcsin}\, x + \mathrm{arccos}\, x = \frac{\pi}{2}$ pre $|x| < 1$;


* $\mathrm{arctg}\, x + \mathrm{arccotg}\, x = \frac{\pi}{2}$


* $\mathrm{arcsin}\, x = \mathrm{arctg}\,\left (\frac{x}{\sqrt{1-x^2}}\right )$ pre $|x| < 1$;


* $\mathrm{arccos}\, x = \mathrm{arccotg}\,\left (\frac{x}{\sqrt{1-x^2}}\right )$ pre $|x| < 1$;


* $\mathrm{arctg}\, x = \mathrm{arcsin}\,\left (\frac{x}{\sqrt{1+x^2}}\right )$; 


* $\mathrm{arccotg}\, x = \mathrm{arccos}\,\left (\frac{x}{\sqrt{1+x^2}}\right )$; 


* $\mathrm{arctg}\, x = \begin{cases}
\ \mathrm{arccotg}\, \frac{1}{x}, \quad x > 0\\
\ -\pi + \mathrm{arccotg}\, x, \quad x < 0
\end{cases}$.

```python
# created by Timothy Clemans
def error_msg(msg):
    pretty_print(html('<p style="font-family:Arial, sans-serif;color:#000"><span style="color:red;font-weight:bold">Error</span>: %s</p>' % msg))

@interact
def interactive_2d_plotter(expression=input_box('sin(x)', 'Expression', str), x_range=range_slider(-10,10,1,(0,10), label='X Range'), square=checkbox(True, 'Square'), axes=checkbox(False, 'Show Axes')):
    if expression:
        try:
            expression = SR(expression) # turn string into a Sage expression
        except TypeError:
            print(error_msg('This is not an expression.'))
            return
        try:
                xmin, xmax = x_range
                if square or not axes:
                    print("var('%s')\nplot(%s).show(%s%s%s)" % (expression.variables()[0], repr(expression), 'aspect_ratio=1' if square else '', ', ' if square and not axes else '', 'axes=False' if not axes else ''))
                    if square:
                        plot(expression, xmin, xmax).show(aspect_ratio=1, axes=axes)
                    else:
                        plot(expression, xmin, xmax).show(axes=axes)
                else:
                    print("var('%s')\nplot(%s)" % (expression.variables()[0], repr(expression)))
                    plot(expression, xmin, xmax).show(axes=axes)
        except ValueError:
            print(error_msg('This expression has more than one variable.'))
            return
        except TypeError:
            print(error_msg("This expression contains an unknown function."))
            return
```

Majme nejakú funkciu $y=f(x)$ a konštantu $C$, potom platia nasledujúce pravidlá: 

* $y=f(x)+C\begin{cases}
\ C>0, \quad \text{posun grafu funkcie } f(x) \text{ o hodnotu } C \text{ nahor,}\\
\ C<0, \quad \text{posun grafu funkcie } f(x) \text{ o hodnotu } C \text{ nadol;}
\end{cases}$ 


* $y=f(x+C)\begin{cases}
\ C>0, \quad \text{posun grafu funkcie } f(x) \text{ o hodnotu } C \text{ vľavo,}\\
\ C<0, \quad \text{posun grafu funkcie } f(x) \text{ o hodnotu } C \text{ vpravo;}
\end{cases}$ 


* $y=Cf(x)\begin{cases}
\ C>1, \quad \text{roztiahnutie grafu funkcie } f(x) \text{ v smere osi } y,\\
\ 0<C<1, \quad \text{stiahnutie grafu funkcie } f(x) \text{ v smere osi } y; 
\end{cases}$ 


* $y=f(Cx)\begin{cases}
\ C>1, \quad \text{stiahnutie grafu funkcie } f(x) \text{ v smere osi } x,\\
\ 0<C<1, \quad \text{roztiahnutie grafu funkcie } f(x) \text{ v smere osi } x; 
\end{cases}$ 


* $y=-f(x) \quad \text{ prevrátenie grafu funkcie } f(x) \text{ súmerne podľa osi } x$; 


* $y=f(-x) \quad \text{ prevrátenie grafu funkcie } f(x) \text{ súmerne podľa osi } y$. 

Transformácie predstavené v predošlom odseku je možné kompaktne zapísať aj takto: 

$$af(b(x+c))+d, $$ 

kde $f(x)$ je nejaká reálna funkcia reálnej premennej, $a,b,c,d$ sú nejaké reálne konštanty. 

* $a$ spôsobuje vertikálne natiahnutie/stiahnutie grafu funkcie $f(x)$: 
    * $|a|>1$ natiahnutie grafu funkcie $f(x)$, 
    * $|a|<1$ zmrštenie grafu funkcie $f(x)$ ,
    * $a<0$ prevrátenie grafu funkcie $f(x)$ "hore nohami" (vzhľadom na os $x$). 
    
    
* $b$ spôsobuje horizontálne natiahnutie/stiahnutie grafu funkcie $f(x)$: 
    * $|b|>1$ stiahnutie grafu funkcie $f(x)$, 
    * $|b|<1$ natiahnutie grafu funkcie $f(x)$, 
    * $b<0$ preklopenie grafu zľava-doprava (podľa osi $y$). 
    

* $c$ spôsobuje horizontálne posunutie grafu funkcie $f(x)$: 
    * $c<0$ posun vpravo, 
    * $c>0$ posun vľavo. 
    
* $d$ spôsobuje vertikálne posunutie grafu funkcie $f(x)$: 
    * $d>0$ posun hore, 
    * $d<0$ posun dole.
