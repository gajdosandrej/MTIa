# <font color=brown> Motivácia </font>

Analytická geometria sa uplatňuje napr. aj pri dizajne videohier či v animáciách. 

Častokrát je potrebné v praktických úlohách poznať **polohu kružnice a priamky**. Predstavme si napr. jednoduchú hru v mobilnom telefóne, kde stíhačka páli laserovou zbraňou (vo forme priamky alebo úsečky) na objekty (napr. asteroidy) v tvare kruhu resp. kružnice (prípadne iné tvary). Ak je potrebné zistiť, či strela zasiahla atakovaný objekt, tak potom je vlastne potrebné poznať polohu strely (priamky alebo úsečky) a atakovaného objektu napr. v tvare gule. 

Táto téma bude predovšetkým venovaná stručnejším poznatkom o **priamkách**, **rovinách**, ich **vzájomných polohách**, **odchýlkach** a pod. Pozrieme sa aj na **analytické vyjadrenie kružnice, gule** a tiež **v akom vzťahu môže byť priamka ku kružnici**. 

***
<a id=ciele></a>
# <font color=brown> Učebné ciele</font>

### Základné 


* Poznať rôzne analytické vyjadrenia základných lineárnych útvarov (priamky, úsečky, polpriamky, roviny) v 2D, 3D a vedieť robiť prevody medzi týmito vyjadreniami. 


* Vedieť v akom vzťahu môžu byť voči sebe priamky, roviny. 


* Poznať a vedieť použiť vzťahy pre odchýlky priamok, rovín, priamky a roviny. 


* Vedieť aplikovať vzťahy pre vzdialenosť bodu od priamky v 2D a pre vzdialenosť bodu od roviny v 3D. 


* Rozumieť analytickému vyjadreniu pre kružnicu i guľu a vedieť v akom vzťahu môže byť priamka voči kružnici. 

### Pokročilé 


* Rozhodnúť, či zadaný bod patrí konkrétnej priamke alebo rovine. 


* Určiť vzájomné polohy priamok, rovín, prípadne nájsť bod/body prieniku. 


* Určiť vzájomnú polohu priamky a kružnice, prípadne nájsť bod/body prieniku. 


* Vypočítať odchýlky priamok, rovín. 


* Nájsť predpis priamky alebo roviny s konkrétnymi vlastnosťami. 


* Určiť vzdialenosť bodu od priamky alebo od roviny, určiť vzdialenosť dvoch rovnobežných priamok alebo rovín.
