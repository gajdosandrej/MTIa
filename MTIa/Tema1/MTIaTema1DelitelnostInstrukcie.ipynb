<a id=motivacia></a>
# <font color=brown> Motivácia</font>

V oblasti **kryptografie (šifrovania)** ale aj v bežnom živote je známa a používaná tzv. **[RSA šifra](https://www.mladyprogramator.sk/rsa-sifra/)**. Ronald Rivest, Adi Shamir a Leonard Adleman v roku 1977 navrhli systém pracujúci s dvoma navzájom závislými kľúčmi – verejným a súkromným. Verejný kľúč sa používa na šifrovanie údajov a súkromný na ich dešifrovanie. Princípom algoritmu RSA je faktorizácia, resp. **rozklad veľmi veľkého celého čísla na súčin dvoch veľkých prvočísel**. Pomocou týchto dvoch prvočísel a verejného kľúča možno ľahko určiť súkromný kľúč. Výpočet súkromného čísla iba z verejného kľúča, bez znalosti prvočísel, je však v súčasnosti výpočtovo nerealizovateľný (kvantové počítače to môžu zmeniť). <br/>
Teoretický základ pre [RSA](https://theconversation.com/the-rsa-algorithm-or-how-to-send-private-love-letters-13191) ale tiež pre **hashovacie funkcie**, časti **numerickej analýzy**, **generátory pseudonáhodných čísel** atď. tvorí oblasť matematiky známa ako **teória čísel**. Jedným zo základných pojmov v teórii čísel je **deliteľnosť**, ktorá prirodzene súvisí s dobre známou operáciou delenia. Deliteľnosť má rôzne zaujímavé vlastnosti využívané v praxi ([ElGamal](https://cs.qwe.wiki/wiki/ElGamal_encryption) šifrovací systém), preto si ich predstavíme. Ďalej sa budeme zaoberať pojmami ako **najväčší spoločný deliteľ**, s čím úzko súvisí **Euklidov algoritmus** (i jeho modifikácia) a nevynecháme ani **najmenší spoločný násobok**, ktorý ako uvidíte, ste už zaiste tiež využívali. Tieto i ďalšie koncepty, s ktorými sa oboznámime neskôr, hrajú dôležitú úlohu napr. aj v [kódovaní](https://melisko.webnode.sk/news/uchovavanie-informacii-kodovanie-/).

1. Prečo nasledujúce tvrdenie vo všeobecnosti neplatí? $\left(\forall a,b\in\mathbb{Z}\right)\,\, a\mid b\Rightarrow |a|\le |b|$. Uveďte príklad - tzn. nájdite konkrétne $a,b$, pre ktoré uvedená implikácia neplatí. Čiže nájdite také konkrétne celé čísla $a,b$, že $a\mid b$ ale súčasne $|a| > |b|$.

2. Vymyslite príklad (tzn. nájdite konkrétne celé čísla $a,b,c,u,v$) pre nasledujúcu vlastnosť deliteľnosti: $\left(\forall a,b,c,u,v\in\mathbb{Z}\right)\left(a\mid b \wedge a\mid c\right)\Rightarrow a\mid (bu\pm cv)$. Na príklade tiež ukážte, že opačná implikácia neplatí, tzn. ak $a\mid (bu + cv)$ nemusí potom platiť, že $a\mid b$ a súčasne $a\mid c$.
